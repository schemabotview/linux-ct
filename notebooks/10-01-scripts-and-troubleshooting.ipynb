## From one-liners to production scripts and troubleshooting

This is the capstone. The first nine modules taught you the *pieces* of a Linux system — the shell, files, permissions, text tools, processes, users, storage, networking, and systemd. This module ties them together into the two skills that actually make a sysadmin: **automating work with scripts**, and **diagnosing a system when something breaks**.

Both skills use the whole map at once:

```
        write it once            ·            fix it fast
   ┌──────────────────────┐          ┌──────────────────────────┐
   │  shell scripting      │          │  troubleshooting          │
   │  variables · if · for │          │  disk · memory · CPU      │
   │  functions · set -e   │          │  network · services       │
   └──────────────────────┘          └──────────────────────────┘
        lx-shell / lx-streams              every subsystem you've met
```

The module has two halves. **First, scripting** — turning the commands you've been typing into repeatable programs: the shebang and arguments, conditionals and loops, functions and exit codes, and the `set -euo pipefail` preamble that makes a script production-safe.

**Second, troubleshooting** — a methodology plus five concrete **playbooks** for the failures you'll actually meet: a full disk, out of memory, a runaway CPU, a dead network, and a service that won't start. Each playbook is a diagnostic flow built from tools you already know — `df`, `du`, `free`, `ps`, `ss`, `journalctl`, `strace` — now aimed at a specific emergency.

We close with the cross-cutting professional layer: **Mandatory Access Control** (SELinux and AppArmor), **performance monitoring** (`vmstat`, `iostat`, `sar`), and a full **LFCS exam** guide — what to expect, how to prepare, and where to go after. By the end you'll have seen the whole system, and know how to build on it *and* fix it.
