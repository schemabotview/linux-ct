## A tour of the filesystem hierarchy

Unlike Windows, Linux has no drive letters. Every disk, device, and file lives under a single tree rooted at **`/`** ("the root").

The top-level directories you'll meet over and over:

| Path | What's there |
|---|---|
| **`/home`** | user home directories — your stuff is in `/home/<you>/` |
| **`/root`** | the root user's home (deliberately *not* under `/home`) |
| **`/etc`** | system configuration — plain-text, editable. "My settings." |
| **`/var`** | things that change as the system runs — logs in `/var/log`, caches |
| **`/tmp`** | temporary scratch, world-writable, usually wiped on reboot |
| **`/usr`** | installed software — `/usr/bin` binaries, `/usr/lib` libraries |
| **`/bin`, `/sbin`** | essential + admin commands (often symlinks into `/usr`) |
| **`/boot`** | the kernel image and bootloader config |
| **`/dev`** | device files (`/dev/sda` = first disk, `/dev/null` = the bin) |
| **`/proc`, `/sys`** | *virtual* dirs exposing process and kernel info (not on disk) |

Don't memorise all of it. Just remember: **config in `/etc`, your stuff in `/home/<you>`, logs in `/var/log`, scratch in `/tmp`.** The rest you'll meet as you need them.
