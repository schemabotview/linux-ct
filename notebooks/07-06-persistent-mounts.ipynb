## Persistent mounts — `/etc/fstab` and UUID

A `mount` command is temporary — gone on reboot. To make a mount permanent, add a line to **`/etc/fstab`** ("filesystems table"). Each line has **six space-separated fields**:

```
UUID=ab12-cd34-...   /mnt/data   ext4   defaults   0   2
   (1) device        (2) point   (3)    (4) opts   (5) (6)
```

1. **device** — what to mount. **Strongly prefer UUID** (below).
2. **mount point** — where (must exist); `none` for swap.
3. **type** — `ext4`, `xfs`, `swap`, `tmpfs`… or `auto` to guess.
4. **options** — `defaults` for normal use; `sw` for swap.
5. **dump** — legacy backup flag; **always `0`** today.
6. **pass** — boot `fsck` order: **`1`** for root, **`2`** for others, **`0`** for none (swap, network).

**Why UUID?** Device names like `/dev/sdb1` can change between boots; a **UUID** is assigned to a filesystem at creation and never changes. See them with **`blkid`** (or `lsblk -f`), and use `UUID=...` (or `LABEL=...`, set with `mkfs -L`) in fstab. Device names should not appear in a production fstab.

⚠️ **Test before rebooting.** A broken fstab drops you into emergency mode at next boot:

```bash
sudo mount -a           # mount everything in fstab not already mounted
findmnt --verify        # validate the syntax
```

The safe pattern: mount manually first to confirm it works, then `umount`, add the fstab line, `sudo mount -a` to test, and only then trust it across a reboot.
