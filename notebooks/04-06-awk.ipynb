## `awk` — the pattern-action language

`awk` is a small programming language for **record-oriented** text. The mental model: for each input line, split it into fields, then run a sequence of `pattern { action }` rules. Its defaults are what make it powerful:

- Input splits into **records** — one per line.
- Each record splits into **fields** on whitespace — **`$1`** first field, **`$NF`** last, **`$0`** the whole line.
- **`NR`** is the record (line) number.

The simplest programs pick out fields:

```bash
awk '{print $1}'          # first field of every line
awk '{print $1, $NF}'     # first and last, space-separated
awk 'NR > 5'              # from line 6 on
awk '/error/'             # lines matching /error/ — like grep
```

Use **`-F`** to change the field separator. The classic is `/etc/passwd` (colon-separated):

```bash
awk -F: '{print $1}' /etc/passwd              # usernames
awk -F: '$3 >= 1000 {print $1}' /etc/passwd   # real users (UID >= 1000)
```

The real power: **`BEGIN{}`/`END{}`** blocks (run before/after all input) plus **associative arrays** for aggregation:

```bash
awk -F: '{shells[$7]++} END {for (s in shells) print shells[s], s}' /etc/passwd
```

That counts how many users have each login shell. **Rule of thumb:** `sed` for simple substitution; **`awk` the moment you have columns, arithmetic, or per-record state.**
