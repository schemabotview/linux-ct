## `grep` — searching for text

`grep` ("global regular expression print") reads input line by line, prints the lines that **match** a pattern, and discards the rest. After `ls`, it's the most-typed command in Linux.

Three shapes:

```bash
grep "error" /var/log/syslog     # search a file
grep "root" /etc/passwd /etc/group   # search several files
ps -ef | grep "bash"             # search another command's output
```

The pattern can be a plain string or a regular expression (next section). Quote it to keep the shell's hands off it.

The flags you'll use constantly:

| Flag | Effect |
|---|---|
| **`-i`** | case-insensitive |
| **`-v`** | invert — print lines that *don't* match |
| **`-n`** | prefix each line with its line number |
| **`-c`** | print only the **count** of matches |
| **`-r`** / **`-R`** | recurse into directories |
| **`-l`** | print only the **filenames** that contain a match |
| **`-w`** | match whole words only |
| **`-o`** | print only the matching part |
| **`-A 3`** / **`-B 3`** / **`-C 3`** | 3 lines **after** / **before** / of **context** |
| **`-F`** | fixed string, not regex (faster, no escaping) |
| **`-E`** | **extended** regex — you almost always want this (next section) |

For searching code, install **`ripgrep`** (`rg`): a modern grep — dramatically faster, respects `.gitignore`, saner regex defaults.
