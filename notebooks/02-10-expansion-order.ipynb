## Expansion order — what the shell does to your line

Before any command runs, the shell rewrites your line through a fixed sequence. Knowing the **order** is what saves you from a whole class of surprising bugs.

1. **History** — `!!`, `!42`, `!ls` become previous commands.
2. **Brace** — `{a,b,c}` → `a b c`; `{1..5}` → `1 2 3 4 5`.
3. **Tilde** — `~` → your home; `~user` → their home.
4. **Variable** — `$name` → its value.
5. **Command substitution** — `$(cmd)` → the command's output.
6. **Arithmetic** — `$(( 2 + 3 ))` → `5`.
7. **Word splitting** — the result is split on whitespace into arguments.
8. **Glob** — `*`, `?`, `[abc]` are matched against filenames.
9. **Quote removal** — surviving `'`, `"`, `\` are stripped before the command runs.

The order explains real bugs. In `name=$(ls *.txt)`, the glob expands first (a file list), `ls` runs, `$(...)` captures its output, and it's assigned to `name`. If those filenames contain spaces, later using `$name` **unquoted** re-splits them on those spaces.

That's the deeper reason quoting is the safe default: **`"$name"` skips word-splitting and globbing** on the value — exactly right when it's meant to be one string.

```bash
echo {a,b,c}.txt          # a.txt b.txt c.txt
echo {1..5}               # 1 2 3 4 5
echo "today is $(date +%A)"
echo $((10 * 7))          # 70
```
