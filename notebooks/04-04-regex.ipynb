## Regular expressions — the brief tour

A **regular expression** (regex) is a tiny language for describing string patterns. `grep`, `sed`, and `awk` all speak it. The full language is large; the part you need for real work is small.

**Two flavours** — this is the first thing to know:

- **Basic (BRE)** — the default in `grep`/`sed`. Metacharacters like `+ ? | ( )` must be backslash-escaped.
- **Extended (ERE)** — enabled with **`grep -E`** / **`sed -E`** (and `awk` uses it by default). They work without escaping.

**Use `-E` by default.** It's what people mean by "regex" today.

The pieces you'll meet most:

| Pattern | Matches |
|---|---|
| **`.`** | any single character |
| **`^`** / **`$`** | start / end of line (anchors) |
| **`*`** / **`+`** / **`?`** | zero-or-more / one-or-more / optional of the previous |
| **`{2,5}`** | between 2 and 5 of the previous |
| **`[abc]`** / **`[a-z]`** | any one of a set / a range |
| **`[^abc]`** | anything **except** the set |
| **`(foo\|bar)`** | `foo` or `bar` (alternation) |

Portable POSIX classes: `[[:digit:]]`, `[[:alpha:]]`, `[[:alnum:]]`, `[[:space:]]`.

Don't swallow the whole language at once. Start with the two **anchors** (`^` `$`), **character classes** (`[abc]`), and **repetition** (`*` `+`) — those cover most real-world greps. The two you'll reach for daily: `^pattern` (lines that *start* with it) and `pattern$` (lines that *end* with it).
