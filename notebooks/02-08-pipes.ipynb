## Pipes — the most important character in the shell

A pipe (`|`) connects the **stdout of one command** to the **stdin of the next**. It's the operator that turns Linux's small, single-purpose tools into something far more powerful than any one of them.

```bash
ls /etc | wc -l
```

`ls /etc` doesn't print to your terminal — its output flows into `wc -l`, which counts lines. The result is the number of entries in `/etc`. Chain as many as you like:

```bash
ls /etc | sort | head -5
ps -ef | grep bash | wc -l
cat /etc/passwd | cut -d: -f1 | sort
```

Each `|` connects stdout on the left to stdin on the right. **stderr is not piped** — it still goes to your terminal unless you redirect it.

This *is* the Unix philosophy: small programs that each do one thing well, read stdin, write stdout, and compose with pipes. You'll lean on it constantly in module 04 (`grep`, `sed`, `awk`, `cut`, `sort`, `uniq`).

One rule to note: a pipeline's exit status is the **last** command's by default. Use `set -o pipefail` (module 10) to make it fail if *any* stage fails.
