## The Unix philosophy

Most systems give you a few big programs that try to do everything — one application that word-processes, checks grammar, does mail-merge, embeds spreadsheets. Linux takes the opposite bet, stated by Doug McIlroy in 1978:

> **Write programs that do one thing and do it well. Write programs to work together. Write programs to handle text streams, because that is a universal interface.**

Every tool in this module is a small program with a narrow job:

- **`grep`** filters lines by pattern · **`sed`** rewrites lines · **`awk`** treats lines as fields
- **`cut`** extracts columns · **`sort`** orders · **`uniq`** collapses adjacent duplicates
- **`tr`** translates characters · **`wc`** counts · **`find`** locates files · **`tar`** bundles them

None is impressive on its own. But the third rule — **text as a universal interface** — is what lets them combine: because every tool speaks lines of text on stdin and stdout, any tool's output can become any other tool's input. Chain them with pipes and you get extraordinarily complex text manipulation in a single line.

That's the payoff of the plumbing you learned in module 02. The "Linux pipeline" is one of the most powerful and underrated programming environments ever built — and after modules 01–03 you already have everything you need to write them. The rest of this module is just learning the individual tools well enough to reach for the right one.
