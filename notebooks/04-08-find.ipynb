## `find` — searching the filesystem

`grep` searches *inside* files; **`find` searches *for* files** — by name, type, size, time, ownership, or any combination. It walks the real tree on every run, so it's slower than `locate`/`fd`, but it's everywhere, even on minimal systems.

The shape is **starting paths**, then a **chain of predicates**:

```bash
find . -type f | wc -l        # count regular files under here
find /etc -name "*.conf"      # config files under /etc
```

The most useful predicates:

| Predicate | Matches |
|---|---|
| **`-name "*.log"`** | basename matches the glob — **quote it!** |
| **`-iname`** | case-insensitive `-name` |
| **`-type f`** / **`d`** / **`l`** | regular file / directory / symlink |
| **`-size +100M`** / **`-1k`** | larger than / smaller than |
| **`-mtime -7`** | modified in the last 7 days (`+30` = older than 30) |
| **`-mmin -10`** | modified in the last 10 minutes |
| **`-user NAME`** / **`-group`** | owned by |
| **`-perm -u+x`** | at least these permission bits |
| **`-empty`** | empty file or directory |
| **`-maxdepth N`** | don't recurse deeper than N |

Combine predicates with **`-a`** (and — the default), **`-o`** (or), **`!`** (not), grouped with `\( ... \)`. Use **`-prune`** to skip whole trees like `node_modules` or `.git`.
