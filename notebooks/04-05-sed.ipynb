## `sed` — the stream editor

`sed` reads input line by line, applies an editing script to each line, and writes to stdout. The overwhelming majority of real `sed` use is **one command** — substitution — so learn that and you've covered ~95% of it:

```
sed 's/PATTERN/REPLACEMENT/FLAGS' file
```

Without flags, only the **first** match on each line is replaced. The **`g`** flag makes it global; **`i`** makes the match case-insensitive:

```bash
echo "one two two" | sed 's/two/TWO/'     # one TWO two   (first only)
echo "one two two" | sed 's/two/TWO/g'    # one TWO TWO   (g = global)
```

As with `grep`, **use `-E`** so you don't backslash-escape `+ ? | ( ) { }`. Parentheses create **capture groups**, and `\1`, `\2` are **back-references** to them — how you reorder text:

```bash
echo "version 1.2.3" | sed -E 's/[0-9]+/X/g'          # version X.X.X
echo "Hello World"   | sed -E 's/(Hello) (World)/\2 \1/'   # World Hello
```

Two more worth knowing:

- **Change the delimiter** when the pattern has slashes (file paths). Any character works: `sed 's|/usr/bin|/opt/bin|g' file`.
- **In-place edit** with **`-i`** — modifies the file instead of printing. `sed -i.bak 's/old/new/g' f` keeps a `.bak` backup. (BSD/macOS `sed` needs `-i ''`.)

When `sed` starts feeling clunky — multiple substitutions, conditionals, anything stateful — reach for `awk`. Use `sed` for what it's great at: simple substitution in a pipeline.
