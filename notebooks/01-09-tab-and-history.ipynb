## Tab completion and history

Two shell features that save more keystrokes than anything else.

**Tab completion.** Start typing a command or path and press **Tab**. If there's one possibility, the shell completes it. Press **Tab twice** to see the list when there are several. Type `ls /et` then Tab → `ls /etc/`. Tab completes commands, paths, filenames, and even some option names. Typing `/usr/share/applications/firefox.desktop` by hand is a mistake — tab through it: fast and typo-proof.

**Command history.** Every command is recorded. Recall with ↑ / ↓. To jump faster:

- **`history`** — show recent commands, numbered.
- **`!!`** — re-run the previous command (great as `sudo !!` when you forgot `sudo`).
- **`!42`** — re-run command number 42.
- **`!ls`** — re-run the most recent command starting with `ls`.
- **Ctrl-R** — interactive reverse search: press it, start typing, it finds the most recent match. Enter runs it; →/Tab edits it first.

History persists across sessions in `~/.bash_history` (last ~1000 commands by default).
