## Jobs — foreground and background

By default the shell runs a command in the **foreground** and waits for it to finish before giving you the prompt back. Append **`&`** to run it in the **background** instead and keep typing:

```bash
sleep 60 &
[1] 12345          # the shell prints [job number] and PID
```

**`jobs`** lists the background jobs of your current shell, each with a **job number** (`%1`, `%2`) that's local to that shell. Move jobs between foreground and background:

```bash
jobs               # list them
fg %1              # bring job 1 to the foreground (bare fg = most recent)
```

The classic "this is taking too long" recovery is a two-step:

1. Press **Ctrl-Z** — the shell **stops** the foreground process (state `T`) and returns the prompt.
2. Type **`bg`** — the stopped job resumes, now in the background.

```
$ long-command
^Z
[1]+  Stopped     long-command
$ bg
[1]+ long-command &
$                  # prompt back, command still running
```

Keep the keys straight: **Ctrl-C** *terminates* the foreground process (`SIGINT`); **Ctrl-Z** merely *pauses* it (`SIGTSTP`). And **`wait`** blocks until your background jobs finish (`wait %1` or `wait <PID>` for a specific one).
