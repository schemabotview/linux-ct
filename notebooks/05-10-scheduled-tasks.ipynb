## Scheduled tasks — `cron`, `at`, systemd timers

Three ways to run something later.

**`cron`** is the classic recurring scheduler. Each user has a **crontab**: `crontab -e` to edit, `-l` to list. A line is **five time fields plus a command**:

```
* * * * *  command
│ │ │ │ └── day of week (0-7, Sun = 0/7)
│ │ │ └──── month (1-12)
│ │ └────── day of month
│ └──────── hour (0-23)
└────────── minute (0-59)
```

Each field takes a number, a range (`9-17`), a list (`1,15`), a step (`*/15`), or `*`. So `30 2 * * *` is daily at 2:30 AM; `0 9 * * 1-5` is 9 AM on weekdays. **The gotchas that bite everyone:** cron runs with a **minimal `$PATH`** (use full paths), it's **non-interactive** (anything that prompts hangs), and **output is lost unless you redirect** (`>> log 2>&1`). Test the script by hand first.

**`at`** schedules a **one-off**: `echo cmd | at now + 1 hour` (`atq` to list, `atrm` to cancel).

**systemd timers** are the modern replacement (module 09): a `.timer` unit paired with a `.service`. They **log to `journalctl`**, catch up missed runs with `Persistent=true`, and list cleanly with `systemctl list-timers`. For LFCS: know **`cron`** deeply, timers at preview level.
