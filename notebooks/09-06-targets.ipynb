## Targets — systemd's runlevels

Old init systems had numbered **runlevels** (0 = halt, 3 = text, 5 = graphical). systemd's equivalent is **targets** — units that *group* other units to define a whole system state.

| Target | Means | Old runlevel |
|---|---|---|
| **`poweroff.target`** | halt and power off | 0 |
| **`rescue.target`** | single-user, root shell, no network | 1 |
| **`multi-user.target`** | multi-user + network, no GUI — **server default** | 3 |
| **`graphical.target`** | multi-user + network + GUI — **desktop default** | 5 |
| **`reboot.target`** | restart | 6 |
| **`emergency.target`** | only `/` mounted, root shell, no services | — |

`default.target` is a symlink to whichever one boots by default. Inspect and switch:

```bash
systemctl get-default                          # what boots by default
sudo systemctl set-default multi-user.target   # change it
sudo systemctl isolate rescue.target           # switch NOW
```

When normal boot is broken, you can reach rescue or emergency mode straight from GRUB by appending **`systemd.unit=rescue.target`** to the kernel command line — a key recovery trick.

**Shutdown and reboot** have modern and traditional spellings that all work:

```bash
sudo systemctl reboot      # or: sudo reboot   /  sudo shutdown -r now
sudo systemctl poweroff    # or: sudo poweroff /  sudo shutdown -h now
sudo shutdown -h +5 "Patching"   # schedule with a broadcast message; -c cancels
```
