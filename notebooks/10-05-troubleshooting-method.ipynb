## Troubleshooting methodology

A sysadmin's real value isn't memorising commands — it's having a **flow** that turns a vague "something's broken" into a specific diagnosis. The same six steps fit almost every problem:

1. **Reproduce.** If you can't reproduce it, you can't confirm a fix. Find the trigger.
2. **Narrow — what changed?** A recent deploy, a new package, a reboot? The best opener is `journalctl --since "1 hour ago" -p err`.
3. **Layer down.** Start at the highest abstraction the user sees and work down: application logs → service status (`systemctl status`) → system journal → kernel ring buffer (`dmesg`).
4. **Bisect time.** When did it last work? `last`, `journalctl -b -1` (previous boot), package history (`apt` logs, `dnf history`).
5. **Isolate variables.** Change one thing at a time. Service won't start? Run its binary by hand with the same environment. Can't reach a host? `ping` its IP directly, bypassing DNS.
6. **Fix, then prevent.** Patch the symptom, then the cause — then add monitoring so *you* see it before users do.

The instinct that ties it together is **isolation**: at every step you're cutting the problem space in half. Is it the app or the system? This host or the network? DNS or the route? Config or permissions? Each answer eliminates half the possibilities.

The next five sections turn this into muscle memory — concrete **playbooks** for the five emergencies you'll meet most: a full disk, out of memory, a runaway CPU, a dead network, and a service that won't start. **Memorise the first command of each** — that's what turns panic into procedure.
