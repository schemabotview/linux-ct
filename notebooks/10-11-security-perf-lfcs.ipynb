## Security, performance, and LFCS prep

Three professional layers close the course.

**Mandatory Access Control (MAC).** The permissions from module 06 are **Discretionary** AC — the owner decides. **MAC** adds a second kernel-enforced policy that constrains *even root*, so a compromised web server can't do everything its user could. Two implementations: **SELinux** (RHEL default — labels every file/process) and **AppArmor** (Ubuntu default — path-based per-program profiles). At LFCS level you don't author policy; you diagnose. The pattern: when DAC looks correct but a service still can't act, **suspect MAC** — flip it to permissive/complain, retry, and if it now works, the policy was the block. SELinux: `getenforce`, `setenforce 0`, `ls -Z`, **`restorecon -Rv`** (the common "wrong context" fix), `ausearch -m AVC`. AppArmor: `aa-status`, `aa-complain`. ⚠️ **Never permanently disable MAC as "the fix"** — use it to *confirm* the cause, then fix the policy.

**Performance monitoring** (the `sysstat` package): **`vmstat 1`** for a one-line CPU/memory/swap/IO pulse — watch **`r`** (run queue > CPU count = CPU-bound), **`wa`** (high = I/O-bound), **`si`/`so`** (swap thrashing). **`iostat -xz 1`** for per-disk detail — **`await`** and **`%util`** are the bottleneck signals. **`sar`** records history for post-incident "what happened at 3am?" analysis.

**The LFCS exam** is 2 hours, hands-on, on a live VM you SSH into — no multiple choice, ~66% to pass. It spans exactly what modules 01–09 covered: essential commands, running systems, users, networking, service config, storage. Tactics that pass it: **read each task fully** (subtle constraints like "use UUID, not device name"), use **`man`** shamelessly (allowed), **make changes persistent** (`enable`, `/etc/fstab`, `--permanent` — graders check across reboot), run inside **`tmux`** so a dropped connection doesn't lose you, and verify your own work (`systemctl status`, `df`, `getent`). Prepare on the *same distro* as the exam, and rehearse each checklist skill until it's under five minutes from a blank prompt.

That's the course. You've walked the whole system — from a shell prompt to the kernel and back — and you now know how to **build on it and fix it.**
