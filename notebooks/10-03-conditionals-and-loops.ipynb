## Conditionals and loops

**`if` runs on a command's exit code** — `0` is true (success), non-zero is false. The condition is *any command*:

```bash
if [[ -f "$config" ]]; then
    echo "config exists"
elif [[ "$count" -gt 10 ]]; then
    echo "too many"
fi
```

Conditions use `test`, which has three forms — `test -f X`, `[ -f X ]`, and the modern bash **`[[ -f X ]]`** (preferred: adds `&&`, `||`, regex, and no need to quote). Operators worth memorising:

- **Files:** `-e` exists · `-f` regular file · `-d` directory · `-r`/`-w`/`-x` readable/writable/executable · `-s` non-empty.
- **Strings:** `-z` empty · `-n` non-empty · `=` / `!=` · `=~` regex.
- **Numbers:** `-eq -ne -lt -le -gt -ge`.

Combine with `&&`/`||`/`!` inside `[[ ]]`, or as short-circuits: `[[ -d /backup ]] || mkdir /backup`. For many branches, **`case`** beats chained `elif` (patterns are globs, `;;` ends each branch).

**Loops** come in three shapes:

```bash
for f in /etc/*.conf; do echo "$f"; done       # over a glob or list
for i in {1..5}; do echo "$i"; done            # over a range

while [[ $count -lt 5 ]]; do ((count++)); done  # while a condition holds
until systemctl is-active nginx; do sleep 2; done   # until it succeeds
```

⚠️ **The canonical way to read a file line by line** is exactly this form — memorise it:

```bash
while IFS= read -r line; do
    echo "Line: $line"
done < /etc/hostname
```

`IFS=` preserves whitespace; `-r` stops backslash mangling. Control flow inside loops: **`break`** exits, **`continue`** skips to the next iteration.
