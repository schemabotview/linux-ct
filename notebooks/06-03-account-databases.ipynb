## Account databases — `passwd`, `shadow`, `group`

Three plain-text files under `/etc` hold the identity data. Learn their fields and most of user management demystifies itself.

**`/etc/passwd`** — the user list, world-readable, one line per user, **seven colon-separated fields**:

```
ubuntu:x:1000:1000:Ubuntu User:/home/ubuntu:/bin/bash
  (1)  (2) (3)  (4)    (5)          (6)         (7)
```

(1) username · (2) password placeholder — always **`x`** today · (3) **UID** · (4) primary **GID** · (5) GECOS (full name / comment) · (6) home directory · (7) login shell. Setting the shell to `/usr/sbin/nologin` forbids login — the standard move for system accounts. Note the hash is *not* here; anyone can read this file.

**`/etc/shadow`** — where password hashes actually live, readable **only by root** (`640`, `root:shadow`) — that's why it's "shadowed" out of the world-readable passwd. Besides the hash it holds **password-aging** fields (last change, min/max age, warning, expiry). The password field can also be **`!`** (locked), **`*`** (never had one — system accounts), or empty (no password — a disaster; never allow it). The hash's `$6$` / `$y$` prefix marks the algorithm (SHA-512 / yescrypt).

**`/etc/group`** — groups and their members, **four fields**: `name : x : GID : member,member`. Your **primary** group comes from passwd field 4; **supplementary** groups list you here. Common ones: **`sudo`**/`wheel` (may `sudo`), `adm` (read logs), `docker` (≈ root — careful).

Don't `grep` these files in scripts — use **`getent passwd alice`** / **`getent group sudo`**, which also consults network directories (LDAP), not just the local files.
