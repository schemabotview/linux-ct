## Managing users — `useradd`, `usermod`, `passwd`, `chage`

All of these need `sudo`. Together they create, modify, and age accounts.

**Create — `useradd`:**

```bash
sudo useradd -m -s /bin/bash -c "Alice Smith" -G sudo,docker alice
```

The flags that matter: **`-m`** create the home directory (forget it and you get a broken account), **`-s`** login shell, **`-c`** full name, **`-G`** supplementary groups, **`-u`** a specific UID. A fresh `useradd` account has **no password (locked)** — set one with `sudo passwd alice`. (Debian's friendlier `adduser` prompts interactively; use `useradd` in scripts.)

**Modify — `usermod`:** same flags, plus rename (`-l`), lock/unlock (`-L`/`-U`), change shell (`-s`). The infamous gotcha:

```bash
sudo usermod -aG docker alice    # ADD to docker  (-a appends)
sudo usermod  -G docker alice    # SET groups — REPLACES all supplementary groups!
```

**Always use `-aG`** to add someone to a group; a bare `-G` silently drops them from every other group.

**Delete — `userdel`:** `userdel alice` keeps the home dir; **`userdel -r alice`** removes home + mail spool. Check nothing's running first (`pgrep -u alice`).

**Passwords & aging.** `passwd` sets a password; `sudo passwd -l/-u alice` locks/unlocks, `-e` expires it (force change next login). **`chage`** controls aging: `chage -l alice` lists it, `-M 90` max age, `-W 14` warn window, `-E 2026-12-31` account expiry, `-d 0` force a change at next login.
