## Logging in — the shell prompt and your session

The first thing you see is a **prompt** — a line ending in `$` or `#`, waiting for you to type:

```
ubuntu@linux-lab:~$
```

Read it left to right:

- **`ubuntu`** — your username.
- **`@`** — a literal separator.
- **`linux-lab`** — the hostname (this machine's name).
- **`~`** — the current directory; `~` is shorthand for your home directory.
- **`$`** — the prompt symbol. `$` means a regular user; `#` means the **root** (administrator) user — be careful when you see it.

The prompt is drawn by the **shell** — the program reading what you type and running it. On most systems the shell is **bash**. When you type a command and press Enter, the shell:

1. Reads your line.
2. Figures out which command you meant.
3. Runs it.
4. Prints any output.
5. Shows a fresh prompt and waits.

That loop — **read, run, print, prompt again** — is your whole conversation with Linux. Everything in this course is a variation on it.
