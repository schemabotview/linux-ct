## Globs — matching filenames with patterns

A **glob** is a wildcard pattern the shell expands into a list of matching filenames **before** your command runs. Type `ls *.txt` and the shell turns `*.txt` into every `.txt` file in the directory, then runs `ls` on that list — `ls` never sees the `*`.

| Pattern | Matches |
|---|---|
| `*` | any run of characters (except a leading `.`) |
| `?` | exactly one character |
| `[abc]` | one character: `a`, `b`, or `c` |
| `[a-z]` | one character in the range |
| `[!abc]` | one character, **not** `a`/`b`/`c` |

Examples: `ls report?.csv` matches `report1.csv` but not `report10.csv`; `ls [Ll]inux*` matches `Linux…` or `linux…`.

Two rules that trip people up:

1. **The shell expands globs, not the command** — by the time `ls` runs, the pattern is already a list of names.
2. **No match → bash passes the literal pattern through.** `ls *.none` becomes `ls *.none`, and `ls` complains. Set `shopt -s nullglob` to make non-matches expand to nothing instead.

**Globs are NOT regular expressions.** `*` in a glob means "any characters"; `*` in a regex means "zero or more of the previous character." Regex comes in module 04.
