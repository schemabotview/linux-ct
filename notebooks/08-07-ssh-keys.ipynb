## SSH keys — the right way to authenticate

Passwords work but are guessable and must be typed every time. **SSH keys** fix both. A key is a cryptographic pair: a **private key** that never leaves your machine, and a **public key** you copy to every server. On connect, the server issues a challenge only the matching private key can answer — **no password crosses the wire**.

**Generate one** — the modern algorithm is **Ed25519** (fast, secure, small):

```bash
ssh-keygen -t ed25519 -C "you@example.com"
```

You get two files: **`~/.ssh/id_ed25519`** (private — never share) and **`~/.ssh/id_ed25519.pub`** (public — safe to copy). **Always set a passphrase** on a real machine; it encrypts the private key on disk, and the agent (next section) means you type it just once per session. For ancient systems that can't do Ed25519, fall back to `ssh-keygen -t rsa -b 3072`.

**Install the public key** on a server with:

```bash
ssh-copy-id user@host        # appends your .pub to the server's ~/.ssh/authorized_keys
```

After that, `ssh user@host` logs you in without a password.

⚠️ **SSH is strict about permissions** and refuses keys if they're too open:

```bash
chmod 700 ~/.ssh
chmod 600 ~/.ssh/id_ed25519 ~/.ssh/authorized_keys
```

The private key and `authorized_keys` must be `600`, the `.ssh` directory `700`. "Permissions are too open" errors are always this.
