## `export` — promoting a variable to the environment

A plain assignment (`name=value`) makes a **shell variable** — only the current shell sees it. To make a variable visible to the programs you launch, **`export`** it:

```bash
MY_VAR="hello"        # shell variable, this shell only
export MY_VAR         # now also in the environment
export MY_VAR="hello" # both in one line
```

You can verify: a shell variable shows up in `set` but not in `env`; an exported one shows up in both.

The key rule is **inheritance flows one way**. When you launch a program, it gets a *copy* of the environment. Changes the child makes never affect the parent. This is why a variable you set *inside a script* doesn't persist after the script exits — the script's shell exited, taking its environment with it.

That's also why shell startup files matter: `~/.bashrc` and `~/.profile` are read when the shell starts and typically **`export`** the variables you want in every session (covered in module 06).

```bash
FOO=local_only
bash -c 'echo "child sees: $FOO"'   # (empty — not exported)
export FOO=now_exported
bash -c 'echo "child sees: $FOO"'   # now_exported
```
