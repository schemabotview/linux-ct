## The seven file types

The **first character** of every `ls -l` line tells you what kind of thing you're looking at. Linux recognises seven:

| Char | Type | What it is |
|---|---|---|
| **`-`** | regular file | text, binaries, images, documents — most files |
| **`d`** | directory | holds other files (really, `name → inode` entries) |
| **`l`** | symbolic link | a pointer to another path |
| **`c`** | character device | unbuffered byte stream — terminals, `/dev/null`, `/dev/random` |
| **`b`** | block device | fixed-size random-access blocks — disks like `/dev/sda`, `/dev/nvme0n1` |
| **`s`** | socket | a communication endpoint between processes — `/run/docker.sock` |
| **`p`** | named pipe (FIFO) | a persistent pipe living on the filesystem |

You'll deal with the first three constantly. The other four are worth recognising:

- **Character and block devices** in `/dev` are how userspace talks to hardware — the kernel does the real work. Reading `/dev/random` gives random bytes; writing to `/dev/null` discards everything.
- **Sockets** are how local services talk — databases, Docker, and system daemons often listen on a socket file in `/run/` or `/tmp/`.
- **Named pipes** are rare day-to-day but show up in some scripted setups.

Don't confuse this with **`file`**: `file` inspects a *regular* file's contents (text vs image vs executable); the first `ls -l` character reports the **type as the kernel sees it**.
