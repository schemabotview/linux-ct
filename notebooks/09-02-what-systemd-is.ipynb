## What systemd is (PID 1)

**systemd** is the **init system** — PID 1 — on every major modern distribution, and also a suite of related daemons (`journald` for logs, `logind` for login sessions, `resolved` for DNS, `timesyncd` for time). It's the first process the kernel starts and the ancestor of everything else (you met PID 1 back in module 05).

What it does:

- **Starts services in dependency order**, in parallel where it can — that's why modern boots are fast.
- **Manages service lifecycles** — restarts crashed services, runs them on a schedule, activates them on demand.
- **Tracks processes with cgroups** — so stopping a service reliably kills *all* its children, no strays left behind.
- **Aggregates logs** into the journal, and **defines boot targets** ("what should be running for a graphical session?").

Everything systemd manages is a **unit**, described by a text file. Where the file lives determines its priority:

| Location | What's there |
|---|---|
| **`/lib/systemd/system/`** | **vendor** units — packages drop their `.service` files here. **Don't edit.** |
| **`/etc/systemd/system/`** | **admin** overrides and your own units. **Higher priority — edit here.** |
| `~/.config/systemd/user/` | per-user units (`systemctl --user`) |

The rule of thumb: you **read** existing units in `/lib/`, and you **write** new ones — or drop-in overrides — in `/etc/`. That split is what lets your changes survive a package update.
