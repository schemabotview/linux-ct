## What Linux is — and the shape of the whole system

Ask ten people what Linux is and you'll get ten answers, all correct. Let's untangle it.

**Linux, strictly, is one program: the kernel.** The kernel is the software that talks directly to your hardware — CPU, memory, disks, network cards. Every time a program wants to read a file or draw to the screen, it ultimately asks the kernel to do it.

But you can't *use* a kernel alone — it has no prompt, no commands, no way in. You need software wrapped around it: a shell to type into, basic commands like `ls` and `cp`, a way to install more. Bundle all that together and you get a **Linux distribution** ("distro") — Ubuntu, Debian, Fedora, RHEL, Arch. They differ in which tools they ship and how they package software, but underneath they run the same kernel.

The mental picture — and it's the map this whole course zooms into:

```
  Your programs (firefox, python, your scripts)
  ──────────────────────────────────────────────
  Userspace tools (ls, cp, bash, systemd)   ← the distribution adds these
  ──────────────────────────────────────────────
  Linux — the kernel                        ← what "Linux" technically means
  ──────────────────────────────────────────────
  Hardware (CPU, RAM, disk, network)
```

**Why it's worth your time:** almost every server on the internet runs Linux. Every Android phone is built on it. The cloud is Linux. Containers are Linux. All 500 of the top supercomputers run Linux. If you work anywhere near software, Linux isn't optional — it's the floor. This course takes you from zero to the **LFCS** (Linux Foundation Certified System Administrator) exam by module 10.
