## The supporting cast — `cut`, `sort`, `uniq`, `tr`, `wc`

The small tools that glue `grep`/`sed`/`awk` together into pipelines.

**`cut`** — extract columns, by delimiter or character position:

```bash
cut -d: -f1 /etc/passwd        # field 1, colon-separated
cut -d, -f2- file.csv          # field 2 to the end
```

Simpler and faster than `awk` when you don't need arithmetic.

**`sort`** — order lines. Default is ASCII, so `10` sorts before `2` — use **`-n`** for numeric. Also **`-r`** reverse, **`-u`** unique, **`-k N`** by field, **`-h`** human sizes (`1K` `2M`).

**`uniq`** — collapse **adjacent** duplicate lines. "Adjacent" is the catch: **almost always preceded by `sort`.** **`-c`** prefixes each line with its count.

**`tr`** — translate or delete single characters (pipe-only, no filename): `tr 'a-z' 'A-Z'` uppercases · `tr -d ',!'` deletes · `tr -s ' '` squeezes repeats.

**`wc`** — count: **`-l`** lines (the one you'll use most), `-w` words, `-c` bytes.

These combine into the **top-N by frequency** idiom — worth memorising:

```bash
... | sort | uniq -c | sort -rn | head
#      group   count    by count   top N
```

```bash
awk -F: '{print $7}' /etc/passwd | sort | uniq -c | sort -rn | head
```

That finds the most common login shells. Swap the front and it answers "top IPs in this log," "most frequent errors," and countless variations.
