## Swap — the memory pressure valve

**Swap** is disk space the kernel uses as overflow when RAM fills. Pages that haven't been touched recently get written out to swap, freeing RAM for active work; when a page is needed again, it's read back in. Swap can be a **dedicated partition** or a plain **file** (more flexible — resize without repartitioning; the modern default).

**Create a swap file:**

```bash
sudo fallocate -l 2G /swapfile     # a 2 GB file
sudo chmod 600 /swapfile           # MUST be root-only
sudo mkswap /swapfile              # format as swap
sudo swapon /swapfile              # enable now
```

Check it with `swapon --show` or `free -h`; disable with `swapoff`. For persistence add to `/etc/fstab`: `/swapfile none swap sw 0 0`.

**Swappiness** (`vm.swappiness`, 0–200, default **60**) controls how eagerly the kernel swaps — lower favours keeping things in RAM:

```bash
sudo sysctl vm.swappiness=10       # temporary; put in /etc/sysctl.conf to persist
```

Rough values: **60** desktop, **10–20** servers (favour RAM), **0** almost never. **Modern note:** with lots of RAM, swap matters less than it did, but you usually still want some — hibernate needs swap ≥ RAM, and a little swap is a pressure-release valve. Cloud VMs often ship with none, so adding a swap file is a common first tweak. (Swap lives in the memory subsystem, not on a filesystem — hence its own map node.)
