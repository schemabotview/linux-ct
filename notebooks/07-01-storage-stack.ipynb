## The storage stack, top to bottom

Before any command, get the picture. When you `cat /etc/passwd`, the bytes travel down through a **stack of layers**, each turning one abstraction into a lower-level one:

```
   files & directories          "/etc/passwd"
   filesystem  (ext4, xfs)       how bytes are organised
   (optional) LVM volume         flexible software volumes
   (optional) RAID array         redundancy across disks
   partition  (/dev/sda2)        a slice of a disk
   block device (/dev/sda)       the kernel's view of the disk
   physical disk (SSD, HDD)      the hardware
```

The point is that the stack is **composable**. A **simple** laptop is just: disk → one partition → ext4 → mounted at `/`. A **production server** stacks more in: several disks → RAID → LVM → many logical volumes → filesystems → mounted at `/`, `/var`, `/home`. The kernel doesn't care which layers you use — **each layer just exposes a "block device" interface to the one above it**, so you mix and match freely.

That gives you a direction for every task:

- **Inspecting?** Work **top-down** — start at a mount point and follow the layers down with `findmnt`, `lsblk`, `mount`.
- **Creating?** Work **bottom-up** — partition the disk, format a filesystem, then mount it.

This module walks the stack in creation order: block devices and how Linux names them, partitions (MBR vs GPT), filesystems, mounting, making mounts persistent with `/etc/fstab`, answering disk-space questions with `df`/`du`, swap, and finally the two "production" layers — **LVM** for flexible volumes and **RAID** for redundancy.
