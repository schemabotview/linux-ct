## Playbook 4 — network is down

**Symptom:** can't reach external sites, slow DNS, SSH dropping. This is the layered flow from module 08 — go through it **in order**, each step assuming the previous passed.

```bash
# 1. Interface + IP?
ip -br link          # any DOWN?
ip -br addr          # any address at all?

# 2. Gateway reachable?
ip route             # what's my gateway?
ping -c3 <gateway>

# 3. Internet by IP? (bypasses DNS)
ping -c3 1.1.1.1

# 4. DNS resolving?
dig +short example.com
dig @8.8.8.8 +short example.com     # if only this works → resolver broken

# 5. Where does the path die?
traceroute -n 1.1.1.1

# 6. Target port reachable?
nc -zv example.com 443

# 7. On the server: am I even listening?
ss -tunlp
```

Each layer eliminates half the problem: no IP is a DHCP/NetworkManager issue; gateway fails but interface is up means the local network; internet-by-IP works but DNS doesn't means the resolver.

**Common immediate fixes:** restart networking (`sudo systemctl restart NetworkManager`); point DNS at a public resolver; ⚠️ if you *suspect the firewall*, disable it only to confirm (`sudo ufw disable`) and **re-enable the moment you're done** — never leave it off as "the fix" (module 08).
