## Files, inodes, and the permission model

Module 01 showed you *where* files live — the filesystem tree, `/etc`, `/home`, `/var`. This module is about *what a file actually is* underneath that tree, and *who is allowed to touch it*.

Here is the surprise that makes everything else click: **a filename is not the file.** On Linux a file is really two separate things.

- The **inode** — a numbered record the filesystem keeps for every file. It holds the file's *metadata* (owner, group, the permission bits, timestamps, size) and pointers to the actual data blocks on disk. The inode does **not** contain the name.
- The **name** — an entry in a directory that maps a human-readable name to an inode number. A directory is just a list of `name → inode` pairs.

Because the name and the data are separate, a lot of Linux behaviour that seems odd at first becomes obvious:

- **Hard links** — two names pointing at the *same* inode. Neither is "the original."
- **Symbolic links** — a tiny file whose contents are *a path to another name*.
- **Renaming and moving** are cheap — you're editing a directory entry, not copying data.
- **Deleting** removes a name; the data only disappears when the *last* name (and no open program) references the inode.

The other half of the module is the **permission model**. Every inode records an owner, a group, and nine permission bits — read / write / execute, for the **user**, the **group**, and **others**:

```
       owner     group     other
        ↓         ↓         ↓
   [ r w x ] [ r w x ] [ r w x ]
```

You see these nine bits every time you run `ls -l`, and you change them with `chmod` and `chown`.

The rest of the module walks this ground in order: creating and viewing files, copying/moving/deleting them, the two kinds of links, the seven file types, and then the permission model end-to-end — reading `ls -l`, `chmod`, ownership, and the `umask` that decides what new files get by default.
