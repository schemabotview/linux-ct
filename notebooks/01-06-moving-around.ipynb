## Moving around — `cd`, `.`, `..`, `~`

To change which directory you're "in," use **`cd`** ("change directory"). A few shorthands save constant typing:

| Shortcut | Means |
|---|---|
| `~` | your home directory (e.g. `/home/ubuntu`) |
| `.` | the current directory |
| `..` | the parent directory (one level up) |
| `-` | the previous directory (toggles back and forth) |
| `/` | the root of the filesystem |

```bash
cd /etc      # go to /etc
cd ..        # up one level → /
cd ~         # (or just `cd`) home
cd -         # toggle back to where you were
```

After a `cd`, `pwd` confirms where you landed.

**Absolute vs relative paths.** A path starting with `/` is **absolute** — measured from the root. A path that doesn't is **relative** — measured from where you are now. From `/var`, `cd log` (relative) takes you to `/var/log`; `cd /log` (absolute) tries a top-level `/log` that probably doesn't exist.
