## TCP/IP refresher

Three ideas let you read almost any networking output.

- **IP address** — identifies a machine on a network. IPv4 is four numbers 0–255 (`192.168.1.42`); IPv6 is eight hex groups (`2001:db8::1`). Every interface gets at least one.
- **Port** — a number 0–65535 identifying a *service* on that machine: **22** SSH, **80** HTTP, **443** HTTPS, **5432** PostgreSQL. **IP + port** uniquely names a network endpoint.
- **Protocol** — how the two ends talk. **TCP** is connection-oriented, reliable, ordered (HTTP, SSH, most services). **UDP** is connectionless, fast, lossy (DNS, DHCP, streaming).

**Addressing in brief.** An IPv4 address is 32 bits. A *network* is a range of addresses, written in **CIDR** notation — `address/prefix-length`:

| CIDR | Meaning |
|---|---|
| `192.168.1.0/24` | 256 addresses — a typical home LAN |
| `10.0.0.0/8` | ~16.7M — corporate / cloud |
| `127.0.0.0/8` | `127.0.0.1` = "localhost", this machine |

The `10.x`, `172.16–31.x`, and `192.168.x` ranges are **private** — never routed on the public internet, so home and corporate networks all reuse them.

A **default gateway** is where packets go when the destination *isn't* on your local network — usually the router. Without one, you can reach local machines but nothing beyond the LAN. For LFCS you don't need to be a network engineer: know the **address / port / protocol** model, that `/24` is a 256-address LAN, and that you need a gateway for anything past it.
