## ACLs — when owner/group/other isn't enough

The three-class model can't express "alice writes, bob reads, everyone else nothing" — there's only *one* owner and *one* group. **POSIX Access Control Lists (ACLs)** fill that gap by granting explicit permissions to **named users and named groups**, layered on top of the standard bits (which still apply).

**Read** them with `getfacl`, **set** them with `setfacl`:

```bash
getfacl file.txt                       # show owner/group/other + any ACL entries
setfacl -m u:bob:r-- file.txt          # grant bob read
setfacl -m g:contractors:rw file.txt   # grant a group read+write
setfacl -x u:bob file.txt              # remove bob's entry
setfacl -b file.txt                    # strip all ACLs, back to plain perms
```

Once a file has ACLs, `ls -l` appends a **`+`** to the permission string — your signal to run `getfacl`:

```
-rw-r--r--+ 1 alice devs 0 file.txt
          ↑ ACL present
```

For directories, a **default ACL** (`-d`) acts as a template inherited by new files inside:

```bash
setfacl -d -m u:bob:rwx /srv/project   # bob gets rwx on anything created here later
```

**Use ACLs for one-off exceptions**, not as your main access-control mechanism — they're harder to audit, easy to miss in backups, and can surprise other admins. Reach for **groups + standard permissions** first. For LFCS, know that ACLs exist, how to read `getfacl`, and basic `setfacl -m`.
