## What a process actually is

A **process** is a running program. The program — a file on disk — is just the recipe; the moment the kernel loads it into memory and executes it, you have a process. The kernel tracks a small bundle of attributes for each one:

- **PID** — a unique number identifying the process while it's alive (reused after it exits, but never shared by two live processes).
- **PPID** — the **parent's** PID: the process that started this one.
- **UID / GID** — the user and group it runs as, which decide what it can read and write.
- **Command** — the program and its arguments.
- **State** — running, sleeping, stopped, or zombie (next sections).

**PID 1 is special.** It's the first process the kernel starts at boot — on modern Linux, almost always **`systemd`** (module 09). Every other process descends from it. If a process's parent dies while the child is still running, the kernel **re-parents** the orphan to PID 1, which adopts it.

Two consequences worth internalising:

- **Every command you run is a new process.** Type `ls /etc` and the shell **forks** a child, the child **execs** `ls`, the shell **waits** for it, then reprints the prompt. A pipeline `a | b | c` creates one process per stage, all running at once.
- **The shell itself is a process.** Log in and a shell process starts; `exit` and it ends. That parent/child, fork/exec/wait rhythm from the hook is happening on every single command.
