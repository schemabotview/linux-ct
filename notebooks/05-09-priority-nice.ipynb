## Priority — `nice` and `renice`

The scheduler shares CPU time among runnable processes; by default everyone gets a fair slice. You tilt the balance with a **nice value**.

**The range is `-20` (highest priority) to `19` (lowest); default `0`.** It's backwards on purpose: a *higher* nice means *lower* priority. The mnemonic is politeness — a nice-19 process is **very willing to yield**; a nice-minus-20 process is selfish about the CPU.

**Only root can lower the nice value (raise priority).** Any user can raise nice (lower priority) on their own processes — so no one can grab more than their share.

```bash
nice -n 10 some-command       # start at nice 10 (lower priority)
renice 5 -p 12345             # change a running process to nice 5
renice 10 -u alice            # renice all of alice's processes
```

There's also **`ionice`** for **I/O** priority, separate from CPU. Class 3 is "idle" — do I/O only when nothing else wants the disk. The classic combo for a backup you don't want to choke production:

```bash
nice -n 19 ionice -c 3 rsync -av source/ dest/
```

For LFCS, know **`nice`** and **`renice`** cold; treat `ionice` as a useful extra.
