## Filesystems — turning blocks into files

A partition is just a slice of bytes. To store files in it you need a **filesystem** — the layout that records the index of files, where each file's data lives, and its permissions. Without one, a partition is empty space.

The filesystems you'll meet:

| Filesystem | Use when |
|---|---|
| **`ext4`** | the default on most distros — stable, mature, general-purpose |
| **`xfs`** | RHEL default; great for large files and parallelism. **Cannot shrink.** |
| **`btrfs`** | snapshots, subvolumes, built-in RAID (openSUSE, Fedora) |
| **`vfat`** / **`exfat`** | universal compatibility (USB, EFI partition) — **no permissions** |
| **`tmpfs`** | RAM-backed — used for `/tmp`, `/run`; vanishes on reboot |

For LFCS, **know `ext4` deeply and `xfs` well**; the rest are situational.

**Make one — `mkfs`**, with the type as a suffix:

```bash
sudo mkfs.ext4 -L data /dev/sdb1     # format as ext4, label "data"
sudo mkfs.xfs /dev/sdb1              # format as XFS
```

⚠️ **`mkfs` destroys everything** on the target — there's no "format but keep data." The target must be unmounted.

**Check/repair — `fsck`.** It runs automatically at boot; you only run it by hand after a crash or suspected corruption:

```bash
sudo fsck /dev/sdb1                  # autodetects type
```

⚠️ **The filesystem must be UNMOUNTED before `fsck`** — running it on a live filesystem can corrupt it. XFS is the exception: `fsck.xfs` does nothing; use **`xfs_repair`** instead.
