## Getting help — `man`, `--help`, `apropos`

You don't memorise every command — you learn how to **look up** what you've forgotten. Linux ships three layers of help, all offline.

**`man <command>`** — the **manual page**, the thorough reference. Inside `man`: **Space** next page, **`/text`** search, **`n`** next match, **`q`** quit. Man pages are numbered by section — `man 5 passwd` is the *file format* of `/etc/passwd`, while `man 1 passwd` is the *command*. Different topics, same name.

**`<command> --help`** — a short summary built into the program itself. Fastest for a quick reminder.

**`apropos <keyword>`** — search every man page's description by keyword — the "I don't know the command, I just know what I want" tool. `apropos compress` lists everything about compression.

The flow when stuck: **`--help`** first (quickest), then **`man`** (most thorough), then **`apropos`** when you don't even know the command's name.

```bash
ls --help | head
man ls
apropos compress
```
