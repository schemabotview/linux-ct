## SSH — the secure shell

**SSH (Secure SHell)** is how you log in to a remote machine and run commands as if you were at its console. It encrypts everything and authenticates both ends, and it's the *universal* admin protocol — identical on every Linux box, cloud provider, and container. Two ends: the **`ssh`** client you run, and the **`sshd`** daemon on the remote, listening on TCP port **22**.

```bash
ssh user@host                 # log in
ssh user@host -p 2222         # custom port  (lowercase -p)
ssh -i ~/.ssh/key user@host   # use a specific key
ssh user@host "uptime"        # run one command and exit
```

**Host key verification.** The first time you connect, SSH shows the host's **fingerprint** and asks you to confirm it:

```
ED25519 key fingerprint is SHA256:abc...xyz.
Are you sure you want to continue connecting (yes/no)?
```

This is the defence against a man-in-the-middle. On that first connection you should **verify the fingerprint out-of-band** (ask the admin, check the cloud console). Once accepted, it's saved in `~/.ssh/known_hosts` and checked silently forever after.

If you later see **`REMOTE HOST IDENTIFICATION HAS CHANGED!`**, the host's key is different from what you stored — usually a reinstalled or moved server, but potentially an attack. **Don't blindly accept it.** Verify the new key out-of-band, then clear the stale entry with `ssh-keygen -R hostname` once you've confirmed the change is legitimate.
