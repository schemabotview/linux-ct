## `su` vs `sudo`

Two ways to act as another user — the difference is **whose password** you type.

**`su` (substitute user)** starts a shell as another user and asks for **that user's password**:

```bash
su - alice          # become alice with a fresh login environment
su -                # become root with root's environment
su -c "cmd"         # run one command as root, then return
```

The **`-`** (login) flag is the one to remember: without it you keep your own environment and directory; with it you simulate a real login — runs the login scripts, moves to their home.

**`sudo` (substitute user do)** runs *one* command as another user (root by default) and asks for **your own password**:

```bash
sudo cmd            # run as root
sudo -u alice cmd   # run as alice
sudo -i             # interactive root shell (like su - root, via sudo's auth)
sudo -k             # forget the cached credential
```

**`sudo` is the modern default**, and the table shows why:

| | `su -` | `sudo` |
|---|---|---|
| whose password? | the target's | **your own** |
| shares root's password? | yes | **no — root can be locked** |
| per-command logging? | no | **yes** (`/var/log/auth.log`) |
| fine-grained allow-list? | no | **yes** (`/etc/sudoers`) |
| time-limited cache? | no | **yes** (~5–15 min) |

On Ubuntu and most cloud images, **root is locked outright** — there is no root password, and all admin flows through per-user `sudo`. One shared root login is hard to audit; per-user policy is easy. Need to become a *non-root* user? `sudo -i -u www-data` does it without sharing any password.
