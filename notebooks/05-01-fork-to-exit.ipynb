## A process from fork to exit

A program is a file on disk — a recipe sitting there, doing nothing. A **process** is what you get the moment the kernel loads that recipe into memory and starts running it. Everything in this module is about those running things: how to see them, control them, signal them, and schedule them.

The one idea to hold onto is the **process lifecycle**, and it's surprisingly small. Almost every process on a Linux system is born the same way:

```
parent ──fork()──▶ child (a copy) ──exec()──▶ new program running
                                                      │
                              runnable ⇄ sleeping ⇄ stopped   (states)
                                                      │
                                    ──exit()──▶ zombie ──parent wait()──▶ gone
```

- **`fork`** — a process clones itself, producing a near-identical child.
- **`exec`** — the child replaces its own memory with a new program (this is how `ls`, `grep`, everything runs).
- The process then lives in the scheduler, moving between **running, sleeping, and stopped**.
- **`exit`** — it finishes, becoming a brief **zombie** until its **parent** collects the exit status with **`wait`**, and then it's gone.

That single loop — fork, exec, run, exit, wait — is the fork-to-exit path this whole module frames. When you type `ls /etc`, your shell forks a child, the child execs `ls`, the kernel schedules it, it prints and exits, and the shell waits and shows you a prompt again.

The rest of the module walks the loop: what a process really is and its identifiers; viewing processes with `ps`, `top`, and `htop`; the five process states; controlling jobs in the foreground and background; keeping work alive after logout; **signals** — the kernel's way of poking a process; sending them with `kill`; tuning priority with `nice`; and finally scheduling processes to run later with `cron` and systemd timers.
