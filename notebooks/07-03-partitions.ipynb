## Partitions — MBR vs GPT

A **partition** is a contiguous slice of a disk with its own boundaries. You partition to put multiple filesystems on one disk (`/`, `/home`, swap), to dual-boot, or just to keep filesystems small.

Partitions are described by a **partition table** at the start of the disk, and there are two formats:

| | **MBR** (Master Boot Record) | **GPT** (GUID Partition Table) |
|---|---|---|
| era | the old standard | the modern standard |
| partitions | 4 primary (or 3 + extended) | 128 by default |
| max disk | **2 TB** | effectively unlimited |
| firmware | legacy BIOS | **required by UEFI** |
| backup table | no | yes, at disk end |

**Use GPT for any new disk** unless you have a specific reason not to. Check which a disk uses with `sudo parted /dev/sda print` (look for "Partition Table: gpt").

Two tools do the partitioning: **`fdisk`** — the classic interactive one (`p` print, `n` new, `d` delete, `t` type, `w` write, `q` quit); and **`parted`** — newer, handles both tables, and has a scriptable non-interactive form:

```bash
sudo parted /dev/sdb mklabel gpt                 # create a GPT table
sudo parted /dev/sdb mkpart primary ext4 0% 50%  # a partition over the first half
```

⚠️ **Any partition change can destroy data.** Back up first, and **triple-check the device path** (`/dev/sda` vs `/dev/sdb`) before writing. Afterward, run **`partprobe`** so the kernel re-reads the table, then `lsblk` to verify.
