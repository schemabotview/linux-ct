## Interfaces and getting an IP

A **network interface** is the kernel's abstraction for a network connection — an Ethernet port, a Wi-Fi card, a VPN or container virtual link. Names you'll meet: **`lo`** (loopback, always `127.0.0.1`), **`eth0`** (legacy) or **`enp0s3`** (modern "predictable" naming), **`wlan0`** (wireless), **`docker0`** (Docker bridge).

The modern unified tool is **`ip`** (from `iproute2`) — the old `ifconfig`/`route`/`netstat` are deprecated:

```bash
ip -br addr          # brief: one line per interface with its address
ip link              # interface state only
sudo ip addr add 192.168.1.50/24 dev eth0   # add an address (temporary — gone at reboot)
sudo ip link set eth0 up                      # bring an interface up
```

An interface needs an address before it can talk, and there are two ways to get one:

- **DHCP** — the interface asks the network and a server hands back an address, mask, gateway, and DNS. Most laptops, VMs, and cloud instances use it.
- **Static** — you set the address by hand. For servers and gear that need a predictable address.

`ip` changes are **temporary**. For *persistent* config, use your distro's tool — usually **NetworkManager**, driven on a headless server by **`nmcli`**. It works on **devices** (the interfaces) and **connections** (config profiles). The one command to know:

```bash
sudo nmcli connection add type ethernet con-name office ifname eth0 \
    ipv4.method manual ipv4.addresses 192.168.1.50/24 \
    ipv4.gateway 192.168.1.1 ipv4.dns "8.8.8.8 1.1.1.1"
sudo nmcli connection up office
```

For DHCP instead, use `ipv4.method auto` and drop the address/gateway/dns fields. (Ubuntu Server often front-ends this with **`netplan`** YAML.)
