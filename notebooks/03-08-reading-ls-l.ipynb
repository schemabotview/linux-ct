## Reading `ls -l`, character by character

A line from `ls -l`:

```
-rw-r--r-- 1 ubuntu ubuntu 1234 Mar 14 10:23 report.txt
```

Read it left to right. The first ten characters are the type plus the nine permission bits, in three groups of three:

```
 -   rw-   r--   r--
 │    │     │     └── other: read only
 │    │     └──────── group: read only
 │    └────────────── owner: read + write, no execute
 └─────────────────── type: - = regular file
```

Then the rest of the line:

- **`1`** — number of hard links to this inode.
- **`ubuntu ubuntu`** — owner name, then group name.
- **`1234`** — size in bytes.
- **`Mar 14 10:23`** — last modification time.
- **`report.txt`** — the name.

So this is a regular file the owner can read and write, everyone else can only read, 1234 bytes, one link.

A directory starts with `d` and typically reads `rwxr-xr-x` — owner does everything, others can read and traverse:

```
drwxr-xr-x 5 ubuntu ubuntu 4096 Mar 14 10:23 projects/
```

An executable shows the `x` bits set: `-rwxr-xr-x` is a program everyone can run. Spend a minute reading `ls -l /etc/` until the ten-character block is instant — it's the most important "ground truth" you'll read in Linux.
