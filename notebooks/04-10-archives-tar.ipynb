## Archives and compression — `tar` and the tar-pipe

Keep two ideas separate:

- **Archive** — bundle many files into one stream. Linux uses **`tar`** ("tape archive").
- **Compress** — make a stream smaller. Several algorithms, chosen by trade-off.

**`tar` has three operations:**

```bash
tar -czf archive.tar.gz dir/    # Create gzipped:  c=create z=gzip f=file
tar -xzf archive.tar.gz         # eXtract
tar -tzf archive.tar.gz         # lisT (peek without extracting)
```

The compression letter picks the algorithm — **`-z`** gzip (`.gz`), **`-j`** bzip2 (`.bz2`), **`-J`** xz (`.xz`), **`--zstd`** zstd (`.zst`). On **extract you can drop it** — modern `tar -xf` autodetects; you only specify when creating.

Which algorithm? For LFCS work, know two: **gzip** — fast, universal, the safe default; **xz** — slow to compress but excellent ratio, for distribution and archival. **zstd** is the fast modern default and increasingly everywhere (systemd journal, Btrfs, recent distros).

**The tar-pipe.** With **`-f -`**, `tar` reads/writes **stdin/stdout** instead of a file — so you can pipe one `tar` into another and copy a tree with no intermediate archive:

```bash
tar -cf - source/ | tar -xf - -C /destination/
```

The classic use is over SSH (module 08): `tar -czf - source/ | ssh host 'tar -xzf - -C /dest'`. For routine machine-to-machine copies you'd usually prefer **`rsync`** (module 08, transfers only changes), but the tar-pipe works on the most minimal system there is. Takeaway: **`-f -` makes `tar` composable** with the rest of the pipeline.
