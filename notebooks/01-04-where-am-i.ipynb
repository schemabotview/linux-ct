## Where am I? — `pwd`, `whoami`, and your first commands

Five tiny commands that tell you who and where you are. Each takes no arguments — type the name and press Enter.

- **`whoami`** — prints your username. Handy after switching users.
- **`hostname`** — prints this machine's name.
- **`pwd`** — "print working directory": where you are in the filesystem right now.
- **`date`** — the system's current date and time.
- **`uname -a`** — kernel info; `-a` means "all".

```bash
whoami
hostname
pwd
date
uname -a
```

These are your orientation tools. Whenever you're unsure who you are or where you are — after an `ssh`, after a `sudo`, after a `cd` you're not sure worked — `whoami` and `pwd` are the two you'll reach for reflexively.
