## Listing a directory — `ls`

A **directory** is the same idea as a folder — a container for files and other directories. The command to list one is **`ls`** ("list"). By itself it lists the current directory; give it a path and it lists that path:

```bash
ls            # current directory
ls /etc       # a specific directory
ls /          # the root of the filesystem
```

The flags you'll use constantly:

- **`-l`** — long format: one file per line, with permissions, owner, size, and time.
- **`-a`** — all: include hidden files (names starting with a dot, like `.bashrc`).
- **`-h`** — human sizes (K/M/G) instead of raw bytes; only useful with `-l`.
- **`-t`** — sort by modification time, newest first.
- **`-r`** — reverse the sort.

Flags combine: `ls -lah` is the same as `ls -l -a -h`. You'll type `ls -la` and `ls -lh` so often they're worth memorising now.

```bash
ls -la /etc | head
ls -lh /var/log
```
