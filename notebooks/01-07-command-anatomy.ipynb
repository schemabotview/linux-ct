## The anatomy of a command

Every Linux command follows the same shape:

```
command   [options]   [arguments]
```

- **`command`** — the program to run (`ls`, `cp`, `grep`).
- **`options`** (flags/switches) — change *how* it behaves. Short form is one letter after `-` (like `-l`); long form is a word after `--` (like `--long`). `ls -a` and `ls --all` are the same.
- **`arguments`** — *what* it operates on: usually files, paths, or text.

Broken down:

```
cp -r /home/ubuntu/work /backup
│  │  └────── two arguments: source and destination
│  └── option: -r (recursive)
└── command: cp
```

Some options take their own value, like `grep --max-count=5 'error' file`. And a lone **`--`** marks the end of options — after it, everything is an argument even if it starts with a dash. That's how you search for a literal dash: `grep -- '-x' file`.

The convention is `command options arguments`, in that order — stick to it for readability.
