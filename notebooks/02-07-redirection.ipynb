## Redirection — pointing the streams (`>`, `<`, here-docs)

By default a command reads from your keyboard and writes to your terminal. **Redirection** re-points those streams — usually to or from a file. Two streams matter: **stdout** (normal output, file descriptor `1`) and **stderr** (errors, descriptor `2`). They're separate so you can redirect one without the other.

**Output:**

| Operator | Effect |
|---|---|
| **`> file`** | write stdout to `file`, overwriting |
| **`>> file`** | append stdout to `file` |
| **`2> file`** | write stderr to `file` |
| **`&> file`** | write both to `file` |
| **`2>&1`** | send stderr wherever stdout is going |
| **`> /dev/null`** | throw output away |

To silence a command entirely: `command > /dev/null 2>&1`. **Order matters** — read redirections left to right.

**Input:**

- **`< file`** — feed a file's contents as input: `wc -l < /etc/passwd`.
- **Here-document** — embed multi-line text; the marker word ends it:

```bash
cat <<EOF
line one
line two
EOF
```

- **Here-string** — a one-liner: `grep root <<< "$(cat /etc/passwd)"`.
