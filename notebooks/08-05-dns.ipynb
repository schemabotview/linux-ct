## DNS — turning names into addresses

You don't connect to `93.184.216.34`; you connect to `example.com`. **DNS** turns the name into the address. The lookup order: **`/etc/hosts`** first, then the system resolver (usually **`systemd-resolved`**), then the configured DNS servers (your ISP's, or public ones like `8.8.8.8` / `1.1.1.1`).

**`/etc/hosts`** is a plain file mapping addresses to names, and it beats every other source — handy for overriding DNS in development or giving a local box a memorable name:

```
192.168.1.10    home-server
```

**`/etc/resolv.conf`** historically listed the `nameserver` entries and a `search` domain for short names. On modern systemd distros it's a **symlink managed by `systemd-resolved`** — don't edit it directly (it gets overwritten); use `resolvectl status` to inspect, or set DNS through NetworkManager (`ipv4.dns`).

To *test* DNS without trusting the local cache, use **`dig`**:

```bash
dig example.com +short          # just the answer
dig @8.8.8.8 example.com        # ask a specific server, bypassing your resolver
dig example.com MX              # a specific record type
```

The key diagnostic: if **`dig @8.8.8.8`** works but plain **`dig`** doesn't, the wider internet is fine — *your resolver* is broken. Record types worth knowing: **A** (IPv4), **AAAA** (IPv6), **CNAME** (alias), **MX** (mail), **NS** (nameservers), **TXT** (SPF/verification), **PTR** (reverse lookup).
