## Create and view files — `touch`, `mkdir`, `cat`, `less`

Two verbs to start with: make a file, then look inside one.

**Make.** `touch <name>` creates an empty file (it was built to update timestamps, but the side effect of creating a missing file is what you'll use it for). `mkdir <name>` makes a directory; add **`-p`** to create a chain of parents in one step and to stay quiet if it already exists.

```bash
touch report.txt a.txt b.txt        # one or many empty files
mkdir projects
mkdir -p projects/linux/notes       # creates every missing parent
```

Two things that catch newcomers: Linux filenames are **case-sensitive** (`Report.txt` ≠ `report.txt`), and spaces need quoting (`touch "my notes.txt"`) — the convention is to avoid them and use dashes or underscores.

**View.** Pick the right tool for the size of the file:

- **`cat <file>`** — dump the whole file to the terminal. Great for small files; painful for a 50,000-line log. (`cat a b` concatenates — that's the name.)
- **`less <file>`** — an interactive pager for anything large. It doesn't load the whole file, so it opens a 10 GB log instantly. Inside: **Space**/**b** to page, **/pattern** then **n** to search, **g**/**G** for start/end, **q** to quit.
- **`head`** / **`tail`** — first or last 10 lines (`-n N` to change the count). The star feature is **`tail -f`**, "follow" mode, which keeps printing new lines as they're appended — the standard way to watch a log: `tail -f /var/log/syslog`.

One more, for *identifying* rather than reading: **`file <path>`** reads the first few bytes and tells you what something actually is — `file /bin/ls` → "ELF 64-bit executable", `file notes.txt` → "ASCII text". Linux doesn't trust extensions.
