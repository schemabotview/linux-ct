## Copy, move, and rename — `cp` and `mv`

**Copy — `cp <source> <destination>`.** If the destination is a directory, the file keeps its name; if it's a new name, it's copied and renamed.

```bash
cp report.txt report-backup.txt      # copy + rename in place
cp report.txt /tmp/                  # copy into a directory (keeps the name)
cp file1 file2 /tmp/                 # many files into a directory
cp -r projects/ /backup/             # a directory needs -r (recursive)
```

Without **`-r`**, copying a directory fails with "omitting directory" — `cp` refuses by default. Handy flags: **`-i`** prompt before overwriting, **`-v`** print each file, **`-p`** preserve timestamps/ownership/permissions, **`-n`** never overwrite, **`-u`** only if the source is newer.

⚠️ `cp` **overwrites silently** by default — if `report-backup.txt` already exists it's gone with no warning. Aliasing `cp` to `cp -i` is a common safety habit.

**Move / rename — `mv`.** On Linux, renaming *is* moving; there's no separate command.

```bash
mv report.txt report-final.txt       # rename in place
mv report.txt /tmp/                  # move into a directory
mv report.txt /tmp/renamed.txt       # move AND rename
```

When source and destination are on the **same filesystem**, `mv` just rewrites a directory entry — nearly instant, no matter the file size (remember: the name is separate from the data). Across filesystems (say, to a USB stick) it falls back to copy-then-delete, which is slower. `mv` shares `cp`'s `-i`/`-v`/`-n`/`-u` flags, but has **no `-r`** — moving a directory always takes its contents.

For copying *between machines* you'll reach for `scp` or `rsync` — that's module 08.
