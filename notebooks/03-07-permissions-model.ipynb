## The permissions model — who can do what

Linux is multi-user from the ground up. Every file's inode records an **owner**, a **group**, and a set of permissions that control what three classes of people can do.

**Three classes — who's asking?**

- **User (u)** — the file's owner.
- **Group (g)** — members of the file's group.
- **Other (o)** — everyone else.

**Three actions — what are they doing?** The meaning shifts between files and directories:

| | on a **file** | on a **directory** |
|---|---|---|
| **read (r)** | view the contents | list the names inside (`ls`) |
| **write (w)** | modify the contents | create / delete / rename entries inside |
| **execute (x)** | run it (program or script) | **enter** it (`cd`) and reach files within |

The directory rules trip people up. **Execute on a directory means "traversable."** A directory you can read but not execute is effectively locked — you can't `cd` into it or reach anything inside. And because deleting a file is *writing to its directory*, you can delete a file you can't write to, as long as you can write to its parent.

Combine three classes with three actions and you get **nine permission bits** per file:

```
       owner     group     other
        ↓         ↓         ↓
   [ r w x ] [ r w x ] [ r w x ]
```

You read these with `ls -l` (next), and change them with `chmod`.
