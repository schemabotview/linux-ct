## Deleting — `rm`, and the careful use of `rm -r`

**`rm <file>` deletes a file. No trash can. No undo. Gone.**

```bash
rm report.txt
rm file1 file2 file3
```

To delete a **directory and everything inside it**, you need **`-r`** (recursive). Add **`-f`** (force) to suppress prompts and remove read-only files:

```bash
rm -r old-project/
rm -rf temp-stuff/
```

`rm -rf` is genuinely dangerous. The infamous `rm -rf /` has ended careers and taken down production servers. Modern `rm` refuses `rm -rf /` by default (`--preserve-root`), but `rm -rf /home/someone/` still runs without a word of warning.

Habits that save you:

- **`rm -i`** asks before each delete — especially worth it with wildcards: `rm -i *.log`.
- **List first, then delete.** Before `rm *.tmp`, run `ls *.tmp` to see exactly what matches. If it looks right, recall the line and swap `ls` for `rm`.
- **Watch variables.** `rm -rf "$dir/"` where `$dir` is empty becomes `rm -rf /`. Always check a variable before deleting through it.
- **`rmdir <dir>`** removes only an **empty** directory — a safer way to say "this should be empty, remove it" than `rm -r`.

Linux has no built-in trash like macOS or Windows. Once `rm` runs, only specialised recovery tools have any chance — and only sometimes.
