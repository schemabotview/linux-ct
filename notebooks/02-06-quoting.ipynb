## Quoting — when the shell takes you literally

`echo Hello   World` prints `Hello World` — **one space**. The shell collapses runs of whitespace between arguments before `echo` ever sees them. To keep them, you **quote**. Quoting tells the shell "treat this literally" — to one of three degrees, each switching off a different amount of interpretation.

**Single quotes** `'...'` — everything is literal. No variable or command substitution.

```bash
echo 'My home is $HOME'    # My home is $HOME
```

**Double quotes** `"..."` — preserve whitespace and most special characters, **but `$var` and `$(...)` still expand**.

```bash
echo "My home is $HOME"    # My home is /home/ubuntu
```

**Backslash** `\` — escapes the single next character. `\$` is a literal dollar sign.

The practical rule: **when in doubt, put double quotes around variable references** — `"$var"`. This is the single most common shell bug for beginners: an unquoted variable containing spaces gets **word-split** into several arguments. Quoting keeps it one piece.

```bash
name="John Smith"
ls "/tmp/$name"    # one argument: /tmp/John Smith
ls /tmp/$name      # TWO arguments: /tmp/John  and  Smith
```
