## The environment — variables that programs inherit

There are actually **two** kinds of variables. **Shell variables** are visible only inside the current shell. **Environment variables** are visible to every program the shell launches — they get *inherited* by child processes when the shell forks and execs them.

The environment is a list of `NAME=VALUE` pairs handed to every command you run. It's how programs discover your home directory, your username, where to find executables, and more. See it all with **`env`** or **`printenv`**; see one with `echo $VAR`.

The ones you'll meet constantly:

| Variable | Holds |
|---|---|
| **`HOME`** | your home directory — what `~` expands to |
| **`USER`** | your username |
| **`PATH`** | colon-separated dirs searched when you type a command |
| **`PWD`** | the current directory (kept current by `cd`) |
| **`EDITOR`** | the editor `git commit`, `crontab -e` launch |
| **`LANG`** | your locale — sort order, date formats |

**`PATH` deserves attention.** When you type `ls`, the shell doesn't know where `ls` lives. It searches each directory in `PATH`, in order, and runs the first match. **`which ls`** shows which one it would run; **`type ls`** is more thorough (also reveals aliases and builtins).
