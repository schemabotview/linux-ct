## Routing — how packets find their destination

When your machine sends a packet to `93.184.216.34`, the kernel consults its **routing table** to decide *which interface* and *which next hop* to use. The table is a list of "if the destination matches this network, use this gateway via this interface," and the kernel picks the **most specific** match.

```bash
ip route            # show the routing table
```

A typical two-line table:

```
default via 192.168.1.1 dev eth0
192.168.1.0/24 dev eth0 src 192.168.1.42
```

- **`default via 192.168.1.1 dev eth0`** — anything not matched by another rule goes to `192.168.1.1` via `eth0`. This is the **default gateway** — the router on a home LAN, a virtual gateway in the cloud.
- **`192.168.1.0/24 dev eth0`** — the local network is *directly attached*, so no gateway is needed; just use `eth0`.

Most machines have exactly these two rules: "local network is reachable directly" plus "everything else goes via the gateway." VPNs and multi-homed servers add more.

Change routes with `ip route add` / `del` / `change` (temporary — persist them through NetworkManager). And the debugging command worth memorising:

```bash
ip route get 8.8.8.8    # which interface + gateway will actually carry this?
```

Use it whenever you're asking "is my traffic really going where I think?"
