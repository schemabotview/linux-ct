## Shell scripting essentials

A **shell script** is just a file of shell commands; the shell reads it line by line and runs each as if you'd typed it. Turning your one-liners into scripts is how work becomes repeatable.

**The shebang.** The first line of an executable script is `#!` followed by the interpreter path — it tells the kernel what to run the file with:

```bash
#!/bin/bash
echo "Hello, $(whoami)!"
```

Make it runnable and call it like a real command:

```bash
chmod +x hello.sh     # one-time: make executable
./hello.sh            # needs +x AND a shebang
```

Common shebangs: **`#!/bin/bash`** (95% of the time), `#!/bin/sh` (portable POSIX, no bash extras), and **`#!/usr/bin/env bash`** (finds bash in `PATH` — works on macOS too).

**Variables** (recap from module 02): `name="Alice"` (no spaces around `=`), double quotes expand `$name`, single quotes don't, and **always quote when reading** — `echo "$name"`.

**Script arguments** arrive as numbered variables — the vocabulary to memorise:

| | Is |
|---|---|
| **`$0`** | script name |
| **`$1` `$2`** | first, second argument |
| **`"$@"`** | all args, each kept as its own word |
| **`$#`** | number of args |
| **`$?`** | exit code of the previous command |

**Always prefer `"$@"`** (quoted) over `$*` when forwarding arguments — it keeps arguments with spaces intact instead of mangling them.
