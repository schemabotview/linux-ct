## Variables — naming values

A **variable** is a label you give a value so you can refer to it later without retyping. Create one with `name=value` — **no spaces around the `=`**. Writing `name = value` is the most common beginner mistake: the shell reads it as a *command* called `name`.

To read it back, put a **`$`** in front: `$name`. The shell substitutes the value.

```bash
greeting="Hello, Linux"
echo $greeting            # Hello, Linux
count=5
echo "I have $count files"
```

Two things to know straight away:

1. **No types.** Everything is a string. `count=5` stores the *string* `"5"`.
2. **Naming convention.** Lowercase for your own variables (`filename`, `user_input`); UPPERCASE for environment variables (`HOME`, `PATH`). Not enforced, but a strong convention.

- **`unset name`** removes a variable (no `$` — you're naming it, not reading it).
- **`set`** lists every variable currently defined (it's long — pipe to `grep`).
