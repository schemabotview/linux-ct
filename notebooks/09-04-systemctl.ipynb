## `systemctl` and reading `status`

**`systemctl`** is the one command for almost every systemd interaction — `systemctl VERB [UNIT]`.

The verbs you'll use daily:

| Verb | Does |
|---|---|
| **`status`** | current state + recent log lines |
| **`start`** / **`stop`** / **`restart`** | now (temporary) |
| **`reload`** | re-read config *without* restarting (if supported) |
| **`enable`** / **`disable`** | start at boot / not |
| **`enable --now`** | enable **and** start in one command |
| **`is-active`** / **`is-enabled`** | scriptable checks |
| **`cat`** / **`edit`** | show the unit file / edit a drop-in |
| **`daemon-reload`** | re-read unit files from disk |

⚠️ **The most important rule: after creating or editing a unit file, run `systemctl daemon-reload`** — systemd caches definitions and won't see your change otherwise. If a change doesn't take, your first two checks are "did I `daemon-reload`?" and "did I `restart`, not just `reload`?"

**Reading `systemctl status`** is the first move when investigating any service:

```
● nginx.service - A high performance web server
     Loaded: loaded (/lib/systemd/system/nginx.service; enabled)
     Active: active (running) since Mon 09:14:22; 2h ago
   Main PID: 1234 (nginx)
```

- **`●`** — green = active, red = failed, white = inactive.
- **`Loaded:`** — where the file is + `enabled`/`disabled` (starts at boot?).
- **`Active:`** — the runtime state: **`active (running)`**, `active (exited)` (a one-shot that finished fine), `inactive (dead)`, or **`failed`**.
- The last log lines print right there — so `failed`? Go straight to `journalctl -u nginx`.
