## Package management — `apt` and `dnf`

Software is installed by a **package manager**: it downloads packages from a **repository**, resolves dependencies, installs cleanly, and can remove later. Two families:

- **Debian / Ubuntu** — `.deb` packages, modern CLI **`apt`** (`dpkg` underneath).
- **RHEL / Fedora / Rocky** — `.rpm` packages, modern CLI **`dnf`** (`rpm` underneath).

The vocabulary maps almost one-to-one:

| Task | apt | dnf |
|---|---|---|
| refresh catalog | `sudo apt update` | *(automatic)* |
| upgrade all | `sudo apt upgrade` | `sudo dnf upgrade` |
| install | `sudo apt install PKG` | `sudo dnf install PKG` |
| remove | `sudo apt remove PKG` | `sudo dnf remove PKG` |
| search / info | `apt search` / `apt show` | `dnf search` / `dnf info` |
| which pkg owns a file | `dpkg -S /path` | `dnf provides /path` |
| clean orphans | `sudo apt autoremove` | `sudo dnf autoremove` |

⚠️ **The `apt update` vs `apt upgrade` confusion:** **`update`** refreshes the *list* of available packages — it installs nothing; **`upgrade`** actually installs newer versions based on that list. **Always `apt update` first**, or you upgrade against a stale catalog. `dnf` refreshes metadata automatically, so it has no separate `update` step (and adds `dnf history` / `dnf history undo N` for rollback, which apt lacks).

When you have a downloaded `.deb`/`.rpm` file directly, prefer the high-level tool so dependencies resolve: `sudo apt install ./pkg.deb` or `sudo dnf install ./pkg.rpm`. The low-level `dpkg -i` / `rpm -i` install but *don't* fetch dependencies (`apt install -f` fixes them afterward).
