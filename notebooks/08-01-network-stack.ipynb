## The network stack and the socket

Every network task on Linux — reading an address, tracing a route, opening an SSH session — comes back to one endpoint: the **socket**. A socket is where a program plugs into the network, and it's identified by a simple triple: **IP address + port + protocol**.

```
   your program  ─┐
                  ▼
        socket = IP : port / protocol      e.g. 93.184.216.34 : 443 / TCP
                  │
   interface (eth0, wlan0) ─▶ routing table ─▶ gateway ─▶ the wider internet
                  │
   name → address (DNS)  ·  who can reach me (firewall)
```

To send a packet, Linux needs four things lined up, and they're exactly the topics of this module:

- an **address** on an **interface** (so the machine is on the network at all),
- a **route** (which interface and gateway a packet should leave through),
- a **name-to-address** lookup so you can say `example.com` instead of a number (**DNS**),
- and a rule about **who's allowed to connect** (the **firewall**).

Once that plumbing works, the rest of the module is about *using* it securely. **SSH** is the universal remote-admin protocol — the same `ssh user@host` on every Linux box, cloud provider, and container — so half the module is SSH end to end: the connection and host keys, key-based auth, the client config and agent, the `sshd` server, and file transfer with `scp`/`rsync`.

We walk it in order: a TCP/IP refresher, interfaces and getting an IP, routing, DNS, then SSH from both ends, and finally firewalls and a layered troubleshooting drill for when the network "isn't working." The modern tools throughout are **`ip`** (addresses and routes) and **`ss`** (sockets) — the old `ifconfig`/`netstat` are deprecated.
