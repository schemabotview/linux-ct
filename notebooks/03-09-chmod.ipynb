## `chmod` — changing permissions

`chmod` has two equivalent syntaxes. Learn to read both — you'll see both in the wild.

**Symbolic form — like a sentence:** `chmod [who][operator][what] file`

- **who**: `u` owner, `g` group, `o` other, `a` all
- **operator**: `+` add, `-` remove, `=` set exactly
- **what**: `r`, `w`, `x`

```bash
chmod u+x script.sh          # add execute for owner
chmod go-w file.txt          # remove write for group + other
chmod u=rw,go=r file.txt     # set exactly: owner rw, group/other r
```

**Numeric (octal) form — one digit per class**, each digit the sum of **read 4 + write 2 + execute 1**:

- `7` = rwx · `6` = rw- · `5` = r-x · `4` = r-- · `0` = ---

Digits go in order **user, group, other**. The combinations you'll see constantly:

| Octal | Symbolic | Used for |
|---|---|---|
| **755** | `rwxr-xr-x` | programs, public dirs — owner all, others read + run |
| **644** | `rw-r--r--` | text/config files — owner edits, others read |
| **700** | `rwx------` | private dirs like `~/.ssh` |
| **600** | `rw-------` | private files like `~/.ssh/id_rsa` |
| **777** | `rwxrwxrwx` | "anyone can do anything" — almost always a security smell |

```bash
chmod 755 script.sh
chmod 600 ~/.ssh/id_rsa
```

Flags: **`-R`** recursive, **`-v`** verbose. ⚠️ A common gotcha: `chmod -R 755 dir/` makes *every regular file* executable. The safe pattern splits files from directories (that's `find`, module 04):

```bash
find dir -type d -exec chmod 755 {} +
find dir -type f -exec chmod 644 {} +
```
