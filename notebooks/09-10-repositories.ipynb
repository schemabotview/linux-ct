## Repositories — where packages come from

Packages come from **repositories** — servers hosting a catalog plus the package files. Your distro ships a default set; you can add more.

**Debian/Ubuntu** — configured in `/etc/apt/sources.list` and `/etc/apt/sources.list.d/*.list`. A line reads:

```
deb http://archive.ubuntu.com/ubuntu jammy main restricted universe multiverse
```

`deb` (binary) · the URL · **`jammy`** the release *codename* · then **components**: **`main`** (supported FOSS), `restricted` (proprietary drivers), **`universe`** (community FOSS), `multiverse` (proprietary).

**RHEL/Fedora** — `.repo` files in `/etc/yum.repos.d/`, INI-style with `baseurl`, `enabled`, `gpgcheck`, `gpgkey`. Manage with `dnf config-manager --enable/--disable`, and list with `dnf repolist`.

⚠️ **GPG signatures gate trust.** Real repositories **sign** their packages, and the package manager verifies the signature before installing — your protection against a tampered mirror. **Never ignore GPG errors**: install the key from a verified source, or don't install the package. Adding a third-party repo (Docker, say) is always: add its GPG key, add the source, refresh, install.

**EPEL** ("Extra Packages for Enterprise Linux") is the practical extra on RHEL-family systems — the default repos are conservative, so `sudo dnf install epel-release` unlocks a lot of common software (like `htop`). It's almost a must-install on fresh RHEL boxes; Ubuntu doesn't need an equivalent since `universe` already covers most of it.

For LFCS: know **where repo config lives** (`/etc/apt/sources.list*` vs `/etc/yum.repos.d/`), how to **add a repo**, and that **GPG keys gate trust**.
