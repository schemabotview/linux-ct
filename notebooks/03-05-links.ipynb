## Links — hard links and symbolic links

Because a name and an inode are separate (§1), one inode can have **more than one name**. That's a link. Linux has two kinds.

**Hard link — `ln <source> <link-name>`.** A second directory entry pointing at the **same inode**. Both names are equally first-class — there's no "original" and "copy." Edit through one name and the other sees it; delete one name and the data survives as long as any name still points to the inode.

```bash
ln original.txt second-name.txt      # both names → the same inode
```

**Symbolic link (symlink) — `ln -s <target> <link-name>`.** A tiny separate file whose *contents are a path* to another name. Read the symlink and the kernel transparently follows it to the target.

```bash
ln -s /etc/hostname host-link        # host-link holds the text "/etc/hostname"
```

| | Hard link | Symbolic link |
|---|---|---|
| what it is | second name for one inode | a pointer holding a path |
| survives target rename? | yes | no — the path goes stale |
| can target a directory? | no | yes |
| cross filesystems? | no | yes |
| in `ls -l` | a regular file (`-`) | a link (`l`) with `→ target` |

**Symlinks are what you'll use 99% of the time**, and they're everywhere: `/bin/sh` → `/bin/bash` (or `dash`), `/usr/bin/python3` → a specific version, a `~/.bashrc` pointing into a dotfiles repo. Spot one in `ls -l` by the `l` type and the `->`; **`readlink -f <symlink>`** resolves a chain to the final target.
