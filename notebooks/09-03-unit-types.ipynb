## Unit types

Every unit file ends in a suffix that declares its **type**. You'll manage a handful constantly and merely recognise the rest:

| Suffix | What it is |
|---|---|
| **`.service`** | a daemon or one-shot process — the one you'll write most |
| **`.socket`** | a socket systemd listens on; starts the matching `.service` on demand |
| **`.timer`** | triggers a unit on a schedule — the modern **cron** replacement |
| **`.target`** | a *group* of units defining a system state — systemd's runlevels |
| **`.mount`** / **`.automount`** | a mount point (an alternative to `/etc/fstab`) |
| **`.path`** | trigger a unit when a file or directory changes |
| **`.device`** / **`.slice`** / **`.swap`** | a hardware device / a cgroup / a swap area |

The three you'll actually author and drive: **`.service`**, **`.timer`**, and occasionally **`.target`**.

List units by type — `list-units` shows what's *loaded right now*, `list-unit-files` shows *every* unit file whether active or not:

```bash
systemctl list-units --type=service          # running services
systemctl list-units --type=service --all    # including stopped
systemctl list-units --type=timer            # active timers
systemctl list-unit-files --type=service     # all service files (enabled or not)
```

Keep the distinction in mind: a *loaded* unit is one systemd currently knows about; a *unit file* is the definition on disk, which may or may not be loaded.
