## Signals — the kernel's notification system

A **signal** is a small message the kernel delivers to a process, each with a number and a name (signal 15 is `SIGTERM`). Signals are how you poke a process asynchronously — *terminate, reload, pause, resume*. When one arrives, the process can **run its default handler** (usually "terminate"), **run a custom handler** the author wrote (e.g. reload config), **ignore** it, or **block** it for later.

Two signals are the exception — they **cannot be caught, blocked, or ignored**; the kernel enforces them:

- **`SIGKILL` (9)** — terminate immediately, no cleanup. The last resort.
- **`SIGSTOP` (19)** — pause immediately. Resume with `SIGCONT`.

`kill -l` lists all ~30. The handful you'll actually use:

| Signal | № | Meaning |
|---|---|---|
| **`SIGTERM`** | 15 | **Polite shutdown** — the default for `kill`. Catchable, so the app can clean up first. |
| **`SIGKILL`** | 9 | **Last resort** — killed instantly, no cleanup. Use when `SIGTERM` is ignored. |
| **`SIGINT`** | 2 | What **Ctrl-C** sends — interactive interrupt. |
| **`SIGHUP`** | 1 | Originally "hangup"; now conventionally **"reload your config"** for daemons. |
| **`SIGTSTP` / `SIGCONT`** | 20 / 18 | What **Ctrl-Z** sends (pause, catchable) / resume a stopped process. |
| **`SIGUSR1` / `SIGUSR2`** | 10 / 12 | Application-defined — often "rotate logs" or "reload." |

The three you'll send constantly: **`SIGTERM`** to ask nicely, **`SIGKILL`** when that's ignored, **`SIGHUP`** to make a daemon reread its config.
