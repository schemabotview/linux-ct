## Sending signals — `kill`, `killall`, `pkill`

Three commands send signals; they differ only in **how you name the target**.

**`kill PID`** — by process ID. Default is `SIGTERM`; pick another with `-NAME` or `-NUMBER`:

```bash
kill 12345           # SIGTERM (polite)
kill -9 12345        # SIGKILL by number — what people usually type
kill -HUP 12345      # tell it to reload config
```

**`killall NAME`** — every process whose executable matches that exact name:

```bash
killall nginx        # SIGTERM to all nginx
killall -9 chrome    # SIGKILL to all chrome
```

**`pkill PATTERN`** — like `killall` but the pattern is a regex, with handy selectors (`-f` whole command line, `-u` by user). Its twin **`pgrep`** returns matching PIDs *without* killing — **always `pgrep` first to check your aim, then swap in `pkill`**:

```bash
pgrep -af long-script.sh    # inspect matches
pkill -f long-script.sh     # then kill them
```

**Escalate politely, don't lead with `-9`.** The standard shutdown is: send **`SIGTERM`**, wait a few seconds for cleanup, and only then **`SIGKILL`** if it's still alive:

```bash
kill -TERM 12345; sleep 5
kill -0 12345 2>/dev/null && kill -KILL 12345   # kill -0 just tests if it's alive
```

`SIGKILL` gives no chance to flush logs, close databases, or release locks — leading with it corrupts state. This escalation is exactly what **`systemctl stop`** does for you (module 09).
