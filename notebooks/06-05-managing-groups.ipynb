## Managing groups

The group-side mirror of the user commands — all need `sudo`.

```bash
sudo groupadd devs               # create
sudo groupadd -g 1500 devs       # ...with a specific GID
sudo groupmod -n developers devs # rename
sudo groupdel developers         # delete (fails if it's anyone's primary group)
```

**`gpasswd`** manages *membership* without touching the rest of a user's account:

```bash
sudo gpasswd -a alice devs       # add alice to devs
sudo gpasswd -d alice devs       # remove alice from devs
sudo gpasswd -M alice,bob devs   # set the whole member list
```

So there are two equivalent ways to add someone to a group — `sudo gpasswd -a alice docker` or `sudo usermod -aG docker alice`. Pick either.

**The gotcha that confuses everyone: new group membership doesn't apply to a running shell.** Groups are loaded at **login**, so an already-open shell won't see the change until the user logs out and back in:

```bash
sudo gpasswd -a alice docker
# alice's current shell still can't use docker...
newgrp docker                    # ...start a new shell with the group active (the workaround)
```

`newgrp` spawns a fresh shell with the named group already active — the quick way to avoid a full logout. Confirm membership any time with `id alice` or `groups alice`.
