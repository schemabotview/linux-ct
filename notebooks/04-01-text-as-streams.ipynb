## Text as streams through pipes

Modules 02 and 03 gave you the two halves this module snaps together. Module 02 taught the shell's plumbing — **stdin, stdout, pipes, redirection**. Module 03 taught you that on Linux almost everything is a file, and most of those files are **plain text**: `/etc/passwd`, config files, log files.

Put those together and you get the idea that defines Linux at the command line: **text is the universal interface, and you process it as a stream flowing through small tools.**

Here's the shape of every pipeline in this module:

```
file ─▶ grep ─▶ sed ─▶ awk ─▶ sort ─▶ uniq ─▶ result
        filter  edit   fields  order   count
```

Each tool reads lines on stdin, does **one** transformation, and writes lines on stdout. None is impressive alone. Chained with `|`, they do text manipulation that would take pages of code in most languages — in a single line.

This module teaches that toolkit in two groups:

- **Text stream tools** — `grep` (filter by pattern), regular expressions (the pattern language), `sed` (rewrite), `awk` (fields and aggregation), and the supporting cast (`cut`, `sort`, `uniq`, `tr`, `wc`) that glue them together.
- **Finding and bundling files** — `find` (locate files by name, type, size, time), `find -exec` (act on what you found), and `tar` (bundle a tree into one stream).

The through-line is the pipe. Every tool here is designed to sit in the middle of a `|` chain, and the real skill is composing them. Module 10 returns to pipelines inside full shell scripts.
