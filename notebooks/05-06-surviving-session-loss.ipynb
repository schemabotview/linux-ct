## Surviving session loss — `nohup`, `disown`, `tmux`

When you log out or your SSH connection drops, the shell sends **`SIGHUP`** ("hangup") to its jobs — and most programs treat that as "time to exit." So by default, **closing the terminal kills your background work.** For anything long-running, you need one of these.

**`nohup`** runs a command with `SIGHUP` ignored, redirecting output to `nohup.out`:

```bash
nohup long-script.sh &      # keeps running after the shell dies
```

**`disown`** removes an *already-running* job from the shell's job table, so it won't get `SIGHUP` on exit:

```bash
some-command &
disown            # shell forgets the most recent job
```

**`setsid`** runs a command in a brand-new session with no controlling terminal — the cleanest "detach completely" primitive.

**The modern answer is a terminal multiplexer — `tmux` (or `screen`).** Instead of detaching each command, you run a persistent session that *hosts* your shells. Your SSH drops, the multiplexer keeps running, you reconnect and **attach** right back where you were:

```bash
tmux              # start a session; work normally
                  # Ctrl-B then D to detach
tmux attach       # reconnect later  (tmux ls to list)
```

For real work, **`tmux` is the right answer** — it solves the whole problem, not one command at a time. `nohup` and `disown` remain handy for quick one-offs.
