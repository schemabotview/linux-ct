## The Linux user model — UID and GID

To the kernel, a "user" is really just a **number** — a UID. Names are a convenience for humans, looked up from a file. There are three rough classes of account, split by UID:

- **`root` — UID 0.** The superuser. Can do anything: read any file, kill any process, ignore permission bits entirely. There's exactly one, and you should **never log in as root for everyday work** — it bypasses every safety rail. Use `sudo` for specific tasks instead.
- **System users — UID ~1–999.** Identities for daemons: `www-data` runs the web server, `postgres` the database. They usually have no password and no login shell — they exist purely so the kernel can control what each service may touch.
- **Regular users — UID 1000+.** Human accounts, each with a home directory, a login shell, and a password. (The 1000 cutoff is set by `UID_MIN` in `/etc/login.defs`; old RHEL started at 500.)

Every user also has a **GID** — a primary group — and can belong to many **supplementary groups**. The kernel checks *all* of them when deciding access.

The principle that holds the model together is the one from the hook: **the kernel checks the running process's UID and GIDs against the file's owner, group, and mode — every single time.** Process credentials plus file metadata equals an allow-or-deny. That's the whole thing; the rest is where those numbers are stored and how you manage them.

Check your own identity:

```bash
id            # uid=1000(alice) gid=1000(alice) groups=1000(alice),27(sudo)...
whoami        # alice
groups        # alice sudo docker
```
