## How Linux starts and stays running

Two questions this module answers: how does a Linux box get from a cold power button to a login prompt, and once it's up, what keeps all its services running? The answer to both is a chain that ends at one program — **systemd**, PID 1.

```
power ─▶ BIOS/UEFI ─▶ GRUB ─▶ kernel + initramfs ─▶ systemd (PID 1)
                                                        │
                    ┌───────────────────────────────────┤
                 services     targets     journal     packages
                (.service)   (runlevels)  (journald)  (apt/dnf)
```

The boot chain hands control down a well-defined sequence: firmware finds a disk, **GRUB** loads the kernel, the kernel uses a tiny **initramfs** to mount the real root filesystem, and then it runs **systemd** as the first process. From that moment on, systemd is in charge of *everything running* — and that's what most of this module is about.

systemd manages the whole system as a set of **units**:

- **`.service`** units are daemons — nginx, sshd, your own app. You control them all with one verb, **`systemctl`**, and read their logs with **`journalctl`**.
- **`.target`** units are systemd's runlevels — "what should be running for a server?" vs "for a graphical desktop?"
- **`.timer`** units are the modern cron.

We'll walk it in a deliberate order: what systemd is and the unit model, unit types, the `systemctl` verb and reading `status`, writing your own service (and overriding a vendor one with a drop-in), targets, and the logging system. Then we drop back down to the **boot chain** in detail (GRUB → initramfs → systemd), and finish with how software gets onto the machine in the first place — **package management** (`apt`/`dnf`) and the **repositories** packages come from.
