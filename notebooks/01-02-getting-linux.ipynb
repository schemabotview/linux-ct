## Getting a Linux to practise on

You learn Linux by typing, not by reading. Here are the common ways to get a prompt, easiest first.

**Cloud VM (most authentic).** Spin up a small server on DigitalOcean, AWS Lightsail, Linode, or Hetzner — a few dollars a month. You SSH in (module 08) and you have a real Linux box you can break and rebuild.

**Multipass (free, local).** Canonical's Multipass gives you a one-command Ubuntu VM on your laptop — free, no signup, macOS/Windows/Linux hosts:

```bash
multipass launch --name linux-lab
multipass shell linux-lab
```

**WSL2 (Windows).** Windows Subsystem for Linux 2 runs a real Linux kernel alongside Windows. Install "Ubuntu" from the Microsoft Store and launch — zero friction, shared files.

**Docker container (instant, disposable).** If you already have Docker:

```bash
docker run -it --rm ubuntu:24.04 bash
```

Fastest way to a prompt, but no systemd by default (limits module 09) and no persistence unless you mount a volume.

**For this course:** pick **Multipass** (on a laptop) or a **cheap cloud VM** (most authentic) and stick with it — switching environments mid-course is a distraction.
