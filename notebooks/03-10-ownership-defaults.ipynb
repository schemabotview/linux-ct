## Ownership and defaults — `chown`, `chgrp`, `umask`

**Ownership.** An inode's owner and group live right alongside the permission bits. Change them with `chown` (owner) and `chgrp` (group), or both at once with `chown owner:group`:

```bash
sudo chown alice file.txt          # owner → alice
sudo chown alice:devs file.txt     # owner → alice, group → devs
sudo chown :devs file.txt          # group only (leading colon)
sudo chown -R $USER:$USER /opt/app/  # give a whole tree back to yourself
```

You almost always need **`sudo`** for `chown` — a regular user can't hand a file to someone else, or security would collapse. This fixes the common case where `sudo`-installed files end up owned by `root`. (Module 06 covers users and groups properly.)

**Defaults — `umask`.** When `touch` makes a file, the kernel picks permissions from a default. The starting maximums are **666** for files (never executable by default) and **777** for directories; your **umask** says which bits to *subtract*.

The usual umask is **022** — "remove write from group and other":

- files: `666 − 022 = 644` → `rw-r--r--`
- dirs: `777 − 022 = 755` → `rwxr-xr-x`

A private default common on servers is **077** (files `600`, dirs `700`) — new files readable only by you. Check it with `umask`; change it for this shell with `umask 077`; make it permanent in `~/.bashrc`. Set the **mask** (what to remove), not the result you want.
