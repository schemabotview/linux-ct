## Mounting — attaching a filesystem to the tree

A filesystem is useless until it's **mounted** somewhere in the directory tree. **`mount`** attaches it at a directory; **`umount`** detaches it:

```bash
sudo mount /dev/sdb1 /mnt/data        # attach at /mnt/data
sudo mount -o ro /dev/sdb1 /mnt/data  # read-only
sudo umount /mnt/data                 # detach (by mount point or device)
```

The mount point must already exist as a directory — `mkdir /mnt/data` first. Anything previously visible there is hidden until you unmount.

Mount **options** worth knowing (comma-separated after `-o`):

| Option | Meaning |
|---|---|
| **`ro`** / **`rw`** | read-only / read-write (default) |
| **`noatime`** | don't update access times on read — a free performance win |
| **`nosuid`** / **`noexec`** / **`nodev`** | ignore SUID bits / block execution / ignore device nodes |
| **`defaults`** | `rw,suid,dev,exec,auto,nouser,async` |

Three flavours of "what to mount": a **real device** (`/dev/sdb1`); a **virtual** kernel filesystem (`proc` on `/proc`, `tmpfs` on `/tmp`); or a **bind mount** (`mount --bind /src /dst`) that makes the same data appear at a second path — heavily used by containers.

To see what's mounted, **`findmnt`** is the friendliest (a clean tree); `mount`, `df -hT`, and `cat /proc/mounts` are the alternatives. These mounts are temporary — they vanish on reboot, which is what `/etc/fstab` (next) fixes.
