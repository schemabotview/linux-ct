## `find -exec` — acting on results

`find` gets its power from **`-exec`**, which runs a command on each match. Two forms — and the difference matters:

- **`-exec CMD {} \;`** — run CMD **once per match**. `{}` is the filename; `\;` ends the command. Forks one process per file → slow for many.
- **`-exec CMD {} +`** — run CMD **once**, appending all matches as arguments. Much faster. **Prefer `+`** unless you truly need per-file execution.

```bash
find . -name "*.tmp" -exec rm {} +          # one rm call for all matches
find . -type d -exec chmod 755 {} +         # dirs → 755
find . -type f -exec chmod 644 {} +         # files → 644
```

That last pair is the **correct** way to do "755 for directories, 644 for files" — a flat `chmod -R 755` would wrongly make every file executable (the module 03 gotcha).

`find` also has a built-in **`-delete`**, safer and faster than `-exec rm`:

```bash
find /tmp -name "*.tmp" -mtime +7 -delete   # temp files older than a week
```

⚠️ Always run the `find` **without** `-delete` first to see what would go.

And **`xargs`** turns stdin into arguments — like `-exec +` for any list-producing command. Use `-print0` with `xargs -0` so filenames with spaces survive:

```bash
find . -name "*.log" -print0 | xargs -0 gzip
```
