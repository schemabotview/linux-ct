## Identity, credentials, and access

Module 03 taught you that every file carries an owner, a group, and permission bits. This module answers the other half of that sentence: **who *are* the owner and the group?** Where do users and groups come from, how do you prove you're one of them, and how does the kernel turn that identity into a yes-or-no on every file you touch?

Linux was multi-user from day one. Hundreds of accounts can share one machine, each isolated from the others, and the whole thing rests on one tiny idea:

```
process credentials              file metadata
  UID · GID · groups     ─┐   ┌─  owner · group · mode bits
                          ▼   ▼
                    ┌───────────────┐
                    │  kernel checks │  ── allow / deny
                    └───────────────┘
```

Every time a process opens a file, the kernel compares the **process's credentials** (its UID, GID, and supplementary groups) against the **file's metadata** (owner, group, permission bits) and makes an access decision. That single check *is* the security model. Everything in this module is the plumbing around it.

The module walks that plumbing in order:

- **Where identity is stored** — the user model (`root`, system, regular), and the databases `/etc/passwd`, `/etc/shadow`, `/etc/group`.
- **Managing it** — creating and aging users, and managing groups.
- **Becoming someone else** — `su` vs `sudo`, and the `/etc/sudoers` policy that makes `sudo` safe.
- **Beyond owner/group/other** — the special bits (SUID, SGID, sticky) and ACLs.
- **Proving who you are** — PAM (the authentication framework) and the shell startup files that set up your session.

Find your own identity any time with **`id`**, **`whoami`**, or **`groups`**.
