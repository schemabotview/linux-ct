## Viewing processes — `ps`, `top`, `htop`

Three tools to see what's running: one snapshot, two live.

**`ps`** gives a **snapshot** — a frozen list of processes right now. It has two historical flag styles that both survive; pick one:

```bash
ps aux      # BSD style: a = all users, u = detailed, x = incl. no-terminal daemons
ps -ef      # SysV style: -e = every process, -f = full format
```

The columns you'll read most: **PID**, **PPID** (parent), **USER**, **%CPU**, **%MEM**, **RSS** (real RAM in KB), **STAT** (state letter — next section), and **COMMAND**. Choose your own columns with `ps -eo pid,ppid,user,%cpu,comm`, and draw the parent/child tree with `ps -ef --forest` or `pstree -p`.

**`top`** is the classic **live** view, refreshing every few seconds, sorted by CPU. Useful keys inside it: **`P`** sort by CPU, **`M`** by memory, **`k`** send a signal, **`u`** filter by user, **`q`** quit. For scripts, run it non-interactively: `top -b -n 1`.

**`htop`** is `top`'s friendlier cousin — coloured, scrollable, mouse-aware, with per-CPU bars and function keys (**F5** tree, **F6** sort, **F9** signal). Not installed by default (`sudo apt install htop`), but strictly better for interactive use. `top` still wins for scripted monitoring via batch mode. (Newer still: **`btop`**, with graphs over time.)
