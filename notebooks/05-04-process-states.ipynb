## Process states — R, S, D, T, Z

In `ps` output the **STAT** column shows one letter for the process's state. Five matter:

| Letter | State | Meaning |
|---|---|---|
| **`R`** | Running | on a CPU, **or runnable** and waiting its turn |
| **`S`** | Sleeping | waiting for an event — responds to signals (most processes) |
| **`D`** | Uninterruptible sleep | waiting on the kernel, usually disk I/O — **can't be killed** |
| **`T`** | Stopped | paused by a signal (Ctrl-Z) or a debugger |
| **`Z`** | Zombie | finished, but the parent hasn't collected its exit status yet |

The intuitions that matter:

- **Almost every process is `S` most of the time.** Sleeping isn't bad — it just means "waiting for input, a timer, a packet." Hundreds of idle sleeping processes is normal.
- **`R` means runnable, not "currently executing."** On a busy CPU many processes sit in `R` waiting their turn.
- **`D` is the dangerous one.** A process stuck in `D` is blocked in the kernel and **can't be killed — not even by `SIGKILL`**. It usually means a hung disk or wedged NFS mount. Persistent `D` → investigate the I/O system.
- **`Z` (zombie) is harmless one at a time** — it's just a PID and an exit code waiting to be collected. Hundreds of zombies under one parent means that parent has a bug (it isn't calling `wait`).

You'll also see modifiers on the letter: **`+`** foreground, **`s`** session leader, **`<`** high priority, **`N`** low priority. So a shell shows as `Ss+` — a sleeping session leader in the foreground.
