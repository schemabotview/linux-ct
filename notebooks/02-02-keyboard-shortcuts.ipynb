## Keyboard shortcuts — the ones that pay back instantly

The shell's command line is editable like a tiny text editor. Beginners arrow-key one character at a time; the shortcuts below move by line and word. They're **Emacs-style**, bash's default editing mode.

| Shortcut | Action |
|---|---|
| **Ctrl-A** / **Ctrl-E** | jump to start / end of line |
| **Ctrl-U** / **Ctrl-K** | delete to start / to end of line |
| **Ctrl-W** | delete the word before the cursor |
| **Ctrl-Y** | paste back what you last deleted (U/K/W) |
| **Alt-B** / **Alt-F** | jump back / forward one word |
| **Ctrl-L** | clear the screen (instant `clear`) |
| **Ctrl-C** | abandon the line / interrupt the running command |
| **Ctrl-D** | end-of-input — on an empty prompt, logs you out |
| **Alt-.** | paste the **last argument** of the previous command |

The magic one is **Alt-.** (Alt + period). After `mkdir /tmp/long/path/to/work`, type `cd ` then Alt-. and it fills in `/tmp/long/path/to/work`. Tap it again to walk back through earlier last-arguments.

Practise these once for five minutes and you'll save hours over a career.
