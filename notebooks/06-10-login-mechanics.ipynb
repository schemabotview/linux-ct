## Login mechanics — PAM and shell startup files

Two pieces finish the module: how the system *authenticates* you, and what runs once it lets you in.

**PAM — Pluggable Authentication Modules.** When you log in, `su`, or `sudo`, the program doesn't check your credentials itself — it calls **PAM**, so authentication becomes *configuration*, not code. Want 2FA, LDAP, or password-complexity rules? Drop in a PAM module. Config lives in **`/etc/pam.d/`**, one file per service (`sshd`, `sudo`, `login`, …), built from four **management groups**:

| Group | Handles |
|---|---|
| **`auth`** | verifying identity — "prove who you are" |
| **`account`** | is the account allowed right now? (expired / locked / time-restricted) |
| **`password`** | password changes — complexity, history |
| **`session`** | session setup/teardown — mount home, log the login, set limits |

You rarely write PAM rules, but you may *read* `/etc/pam.d/sshd` to explain a denied login. ⚠️ A PAM mistake can lock you out — always keep a root session open while editing.

**Shell startup files** set up your environment once you're in, and *which* runs depends on the shell type:

- **Login shell** (TTY, SSH, `su - alice`) → `/etc/profile` → `~/.bash_profile` (or `~/.profile`).
- **Interactive non-login** (new terminal, `bash`) → `/etc/bash.bashrc` → **`~/.bashrc`**.

Convention: `~/.bashrc` holds aliases, functions, and your prompt; `~/.bash_profile` holds `export`s and `PATH` — and usually just sources `~/.bashrc` so it runs in both cases. These are the dotfiles hinted at in module 03; after editing, `source ~/.bashrc` to apply.
