## Block devices and naming — `lsblk`

A **block device** is the kernel's abstraction for any storage — a disk, SSD, USB stick, RAID array, or LVM volume — anything that reads and writes in fixed-size blocks. They all appear as files under `/dev/`.

The naming patterns you'll meet:

| Pattern | What it is |
|---|---|
| **`/dev/sda`, `sdb`** | SATA / SCSI / USB disks — `a`, `b` for first, second disk |
| **`/dev/sda1`, `sda2`** | partitions on `sda` (the number is the index) |
| **`/dev/nvme0n1`** | first NVMe drive, first namespace (modern SSDs) |
| **`/dev/nvme0n1p1`** | a partition on NVMe — note the **`p`** before the number |
| **`/dev/vda`** | virtio disk — common inside VMs and cloud instances |
| **`/dev/mapper/<name>`** | device-mapper targets — LVM volumes, LUKS encryption |
| **`/dev/md0`** | software RAID arrays (`mdadm`) |

⚠️ **Device names can change between boots.** Unplug the second disk and reboot, and what was `/dev/sdb` may become `/dev/sda`. **Never reference a disk by device name in `/etc/fstab`** — use UUID or LABEL (§6).

The first command to run on any unfamiliar system is **`lsblk`** ("list block devices") — it prints the whole tree of disks, partitions, and where each is mounted:

```
sda           100G  disk
├─sda1        512M  part  /boot/efi
└─sda3       98.5G  part
  └─ubuntu-root    lvm   /
```

Add **`-f`** to also show filesystem type, label, and UUID. Two companions: **`blkid`** prints UUID + type for a device, and **`findmnt`** shows mounts as a clean tree.
