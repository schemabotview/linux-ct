## Special permission bits — SUID, SGID, sticky

Beyond the nine `rwx` bits from module 03 there are **three special bits**, and they show up as modifiers in the `x` positions of `ls -l`.

**SUID (Set User ID) — `4000`.** On an executable, the program runs **as the file's owner**, not the user who launched it. The canonical case is **`/usr/bin/passwd`**: you need to write your hash into root-owned `/etc/shadow`, so `passwd` is owned by root with SUID set — running it briefly makes *you* root, just for that program. It shows as **`s`** in the owner-execute slot: `-rw**s**r-xr-x`. Set with `chmod u+s` or `chmod 4755`. **SUID is dangerous** — a bug in a SUID-root binary is a privilege-escalation hole, so systems keep the list tiny. Audit it: `find / -perm -4000 -type f 2>/dev/null`.

**SGID (Set Group ID) — `2000`.** On an executable, runs with the file's group (rare). Far more useful **on a directory**: new files inside **inherit the directory's group** instead of the creator's — the standard pattern for shared project dirs. Shows as `s` in the group slot:

```bash
sudo chgrp devs /srv/project
sudo chmod 2775 /srv/project     # 2 = SGID → all new files owned by group devs
```

**Sticky bit — `1000`.** On a shared-writable directory, only a file's **owner (or root) can delete or rename it** — even though everyone can create files. The classic is **`/tmp`** (`drwxrwxrwt`), so you can't `rm` someone else's temp files. Set with `chmod +t` or `chmod 1777`.

Numerically, prefix the usual three digits with one more: **4** SUID, **2** SGID, **1** sticky — e.g. `chmod 2775 /srv/project`, `chmod 1777 /tmp`.
