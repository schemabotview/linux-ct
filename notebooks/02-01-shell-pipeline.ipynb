## The shell as a machine — line in, program out

In module 01 you met the shell's loop: **read, run, print, prompt**. This module opens the "run" step and shows the machine inside it. Every line you type is processed the same way, and it's worth seeing the whole path once before the details.

When you press Enter, the shell doesn't just hand your line to a program. First it **rewrites** the line through a fixed sequence of expansions — history (`!!`), braces (`{a,b}`), the tilde (`~`), variables (`$HOME`), command substitution (`$(...)`), arithmetic (`$((...))`), then word-splitting and filename globbing (`*.txt`). Only the *result* of all that becomes the command and its arguments.

Then the shell runs it. For a builtin like `cd` it acts directly; for an external command like `ls` it asks the kernel — through the C library — to **`fork`** a child and **`exec`** the program in it, wiring up the three standard streams (**stdin**, **stdout**, **stderr**) so input and output flow where you asked. Redirection (`>`) and pipes (`|`) are just the shell rearranging those streams before the program starts.

So the shell sits at the very top of the stack, translating your text into kernel requests:

```
your line ─▶ expansion (8 phases) ─▶ command + args
                                      │
                              fork + exec  (via libc → syscall)
                                      │
                          a running program, streams wired
```

The rest of this module walks that path: shortcuts to edit the line, variables and the environment it carries, quoting to control expansion, and the stream plumbing — redirection and pipes.
